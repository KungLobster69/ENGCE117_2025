```python
#Level 1

def super_sum(*nums):
    total = 0
    for number in nums:
        total += number
    return total

#------------------------------------------------------------------------------

print(super_sum(1,2,3,4,5))
print(super_sum(6,7,8,9,10))
```

```python
#Level 2

def flexi_calc(mode, *args):
    if mode == "add":
        return sum(args)
    elif mode == "mult":
        total = 1
        for num in args :
            total *= num
        return total
    else:
        return "Error: Invalid mode"
    
# Test Addition
print(flexi_calc("add", 10, 20, 30))      # Output: 60

# Test Multiplication
print(flexi_calc("mult", 2, 3, 4))        # Output: 24 (2 * 3 * 4)

# Test Mixed/Zero (Edge cases)
print(flexi_calc("mult", 5, 0, 10))       # Output: 0
print(flexi_calc("unknown", 1, 2))        # Output: Error: Invalid mode
```

```python
#Level 3

def stat_calc(mode, *args):
    if len(args) == 0 : return 0
    if mode == "avg": return sum(args) / len(args)
    elif mode == "max": return max(args)

print(stat_calc("avg",10,20,30))
print(stat_calc("max", 5, 100, 2))
print(stat_calc("avg"))
print(stat_calc("max"))
```
