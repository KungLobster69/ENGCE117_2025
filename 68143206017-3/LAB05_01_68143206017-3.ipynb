```python
##Level 1 (Easy)

class Robot:
    pass

r1 = Robot()

print(r1)
```

```python
##Level 2 (Medium)

class Robot:
        
    def __init__(self, name, model, battery):
        self.name = name
        self.model = model
        self.battery = battery

r1 = Robot("Commander","V1",100)
print(f'Name: {r1.name}')
print(f'model: {r1.model}')
print(f'Battery {r1.battery}%')
```

```python
##Level 3 (Hard)

class Robot:
        
    def __init__(self, name, model):
        self.name = name
        self.model = model

r1 = Robot("Commander","V1")
r2 = Robot("Commander","V1")

is_same = (r1 is r2)

print(f'Is r1 the same as r2? {is_same}')
print(f'Address r1: {id(r1)}')
print(f'Address r2: {id(r2)}')


```
