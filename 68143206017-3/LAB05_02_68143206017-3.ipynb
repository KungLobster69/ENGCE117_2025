```python
##Level 1 (Easy)

class Robot:
        
    def __init__(self, name, model):
        self.name = name
        self.model = model

r1 = Robot("R-01","Explorer")
print(f'Robot:{r1.name}({r1.model})')
```

```python
##Level 2 (Medium)

class Robot:
        
    def __init__(self, name, battery=100):
        self.name = name
        self.battery = 100
    
r2 = Robot("R-02")

print(f'{r2.name} Battery:{r2.battery}%')

```

```python
##Level 3 (Hard)

class Robot:
        
    def __init__(self, name, battery=100):
        self.name = name
        self.battery = battery
    
        if self.battery < 20:
            print(f"คำเตื่อน:{self.name} มีแบตเตอรี่ต่ำกว่า 20%")

r3 = Robot("R-03",15)

r4 = Robot("R-04")

r5 = Robot("R-05",10)
```
