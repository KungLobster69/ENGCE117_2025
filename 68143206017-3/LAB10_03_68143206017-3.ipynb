```python
##WorkShop03
import tkinter as tk

def submit():
    if accept_var.get():
        print(f"Username: {username_var.get()}")
    else:
        print("Please accept the terms before submitting.")

root = tk.Tk()
root.title("Registration Form")
root.geometry("300x200")

username_var = tk.StringVar()
accept_var = tk.BooleanVar()

tk.Label(root, text="Username:").pack(pady=(10, 0))

username_entry = tk.Entry(root, textvariable=username_var)
username_entry.pack(pady=5)

check_btn = tk.Checkbutton(root, text="I accept the terms", variable=accept_var)
check_btn.pack(pady=5)

submit_btn = tk.Button(root, text="Submit", command=submit)
submit_btn.pack(pady=10)

root.mainloop()
```
