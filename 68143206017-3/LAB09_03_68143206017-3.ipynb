```python
##WorkShop01
def to_upper(func):
    def wrapper():
        result = func()
        
        if isinstance(result, str):
            return result.upper()
        return result  
    return wrapper



@to_upper
def get_hello():
    return "hello world"

@to_upper
def get_greeting():
    return "nice to meet you"


print(get_hello())      
print(get_greeting())   
```

```python
##WorkShop02
def to_upper(func):
    def wrapper():
        result = func()
        return result.upper() if isinstance(result, str) else result
    return wrapper

def add_brackets(func):
    def wrapper():
        result = func()
        
        if isinstance(result, str):
            return f"[{result}]"
        return result
        
    return wrapper


@add_brackets
@to_upper
def get_message():
    return "python is fun"

print(get_message()) 
```

```python
##WorkShop03
def to_upper(func):
    def wrapper():
        return func().upper()
    return wrapper

# Decorator ตัวที่ 2
def add_brackets(func):
    def wrapper():
        return f"[{func()}]"
    return wrapper

@add_brackets    
@to_upper      
def get_text():
    return "hello python"

print(get_text()) 
```
