```python
##WorkShop02
import tkinter as tk

def change_text():
    label.config(text="คลิกแล้วจ้า!", fg="red")

root = tk.Tk()

root.geometry("200x150")

label = tk.Label(root, text="ยังไม่ได้คลิก")
label.pack(pady=10)

button = tk.Button(root, text="คลิกฉันสิ!", command=change_text)
button.pack(pady=10)

root.mainloop()
```
