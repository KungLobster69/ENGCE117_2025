```python
#Level 1

raw_names = [" somchai", "alice", "BOB", "c"]

clean_names = list(map(lambda x: x.strip().lower(), raw_names))

print(clean_names)

final_names = list(filter(lambda x: len(x) > 1, clean_names))

print(final_names)
```

```python
#Level 2

clean_names = ['somchai', 'alice', 'bob', 'c']

final_names = list(filter(lambda x: len(x) > 2, clean_names))

print(final_names)
```

```python
#Level 3

raw_names = [" somchai ", "alice", " BOB", "c"]

# One-liner solution using List Comprehension
names = [n.strip().lower() for n in raw_names if len(n.strip()) > 1]

print(names)
```
