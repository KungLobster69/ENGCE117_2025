```python
##WorkShop02
class Playlist:
    def __init__(self, songs):
        self.songs = songs

    def __len__(self):
        return len(self.songs)
    
    def __getitem__(self, index):
        return self.songs[index]
    
    def __setitem__(self, index, new_song):
        self.songs[index] = new_song

    def __contains__(self, song_name):
        return song_name in self.songs
    
my_playlist = Playlist(["KillerQueen" , "Metalica" , "TEST ME"])

my_playlist[0] = "Supernova"
print(f"Edit index 0:{my_playlist[0]}")

search_song = "KillerQueen"
if search_song in my_playlist:
    print(f"Have '{search_song}' in Playlist!")
else:
    print(f"Don't Know '{search_song}'")


print(f"Have a 'KillerQueen' in Playlist?: {'KillerQueen' in my_playlist}")


```

```python
##WorkShop03
class Playlist:
    def __init__(self, songs):
        self.songs = songs

    def __len__(self):
        return len(self.songs)

    def __getitem__(self, index):
        if isinstance(index, slice):
            sliced_songs = self.songs[index]
            return Playlist(sliced_songs)
        
        return self.songs[index]

    def __setitem__(self, index, new_song):
        self.songs[index] = new_song

    def __contains__(self, song_name):
        return song_name in self.songs

    def __repr__(self):
        return f"Playlist({self.songs})"

p = Playlist(["Seven", "Magnetic", "Cruel Summer", "Supernova", "Hype Boy"])

sub_playlist = p[0:2]

print(f"Type: {type(sub_playlist)}")
print(f"Songs in Playlist news: {sub_playlist}") 

print(f"Total songs in otherLists: {len(sub_playlist)}")
```
