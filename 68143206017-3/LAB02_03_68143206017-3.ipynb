```python
#Level 1
inventory = {'apple':10,'banana':20}

def check_price(item_name):
    return inventory.get(item_name, 0)

item_to_check = 'durian'
qty = check_price(item_to_check)

print(f"Check {item_to_check.capitalize()}:{qty}")
```

```python
#Level 2
inventory = {'apple':10,'banana':20}

def add_stock(item, quantity):
    if item in inventory:
        inventory[item] = inventory[item] + quantity
    else:
        inventory[item] = quantity

add_stock('apple',5)
print(inventory)
```

```python
#Level 3
inventory = {'apple':10,'banana':20}
brunch2 = {'banana':5, 'orange':10}

for item, quantity in brunch2.items():
    if item in inventory:
        inventory[item] += quantity
    else:
        inventory[item] = quantity

print(inventory)
```
