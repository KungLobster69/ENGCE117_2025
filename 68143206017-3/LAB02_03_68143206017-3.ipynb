```python
#Level 1

try:
    with open("note.txt", "w") as file:
        # Write the text into the file
        file.write("Nickname: Prathom")
except FileExistsError:
    print("Error:File already exists")
```

```python
#Level 2

with open("note.txt", "w", encoding="utf-8") as file:
    file.write("Nickname: Prathom")
    
print("File created and nickname written.")

food = input("Please enter your favorite food: ")

with open("note.txt", "a", encoding="utf-8") as file:
    
    file.write("\nFavorite Food: " + food)

print("Favorite food appended to the file.")
```

```python
#Level 3

try:
    with open("note.txt", "r", encoding="utf-8") as file:
        content = file.read()  
        print(content)         
except FileNotFoundError:
    print("Error: The file 'note.txt' was not found. Please run Level 1 first!")
```
