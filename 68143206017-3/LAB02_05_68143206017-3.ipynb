```python
#Level 1
names = ["alice", "bob", "charlie"]

uppernames = [name.upper() for name in names]

print(f"Upper:{uppernames}")
```

```python
#Level 2
prices = [100, 250, 80, 500, 120]

vat_prices = [price * 1.07 for price in prices if price > 200]

print(f"VAT Prices:{vat_prices}")
```

```python
#Level 3
currencises = {'USD':35, 'EUR':38, 'JPY':0.25}

inverted_currencises = {value: key for key, value in currencises.items() if value > 1}

print(f"Inverted:{inverted_currencises}")
```
