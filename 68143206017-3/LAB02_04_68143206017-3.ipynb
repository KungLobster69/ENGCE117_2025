```python
#Level 1
visitors = ['user1','user2','user1','user3','user2']
visitors_set = (set(visitors))
unique_visitors = set(visitors_set)
unique_count = len(unique_visitors)
print(f"Unique Count:{unique_count}")
print(f"Unique Visitors:{unique_visitors}")
```

```python
#Level 2
group_a = {'Reading', 'Coding', 'Gaming'}
group_b = {'Coding', 'Traveling', 'Gaming'}

print(f"Common:{group_a & group_b}")
```

```python
#Level 3
group_a = {'Reading', 'Coding', 'Gaming'}
group_b = {'Coding', 'Traveling', 'Gaming'}

print(f"Only A:{group_a - group_b}")
print(f"Diff:{group_a ^ group_b}")
```
