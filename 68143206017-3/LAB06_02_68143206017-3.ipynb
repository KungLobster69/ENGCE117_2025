```python
class Square:
    def __init__(self, side):
        self.side = side 

    @property
    def side(self):
        return self.__side

    @side.setter
    def side(self, value):
        if value > 0:
            self.__side = value
        else:
            print("คำเตือน: ความยาวด้านต้องมากกว่า 0!")
            
    @property
    def area(self):
        return self.__side * self.__side
    
box = Square(4)
print(f"ความยาวด้าน: {box.side}") 
print(f"พื้นที่: {box.area}")     

box.side = 10  
box.side = -2  
```

```python
try:
    s = Square(4)
    print(f"Side: {s.side}, Area: {s.area}") 

    s.side = 5
    print(f"New Area: {s.area}")           

    s.side = -10 

except ValueError as e:
    print(e) 

try:
    s.area = 100
except AttributeError:
    print("Error: AttributeError (Read-only)") 
```
