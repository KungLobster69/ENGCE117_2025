```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def draw(self):
        pass

class Circle(Shape):
    def draw(self):
        print("ðŸ”´")
        pass
            
class Square(Shape):
    def draw(self):
        print("ðŸŸ¦")
        pass

class Triangle(Shape):
    def draw(self):
        print("ðŸ”º")
        pass    

def render_screen(shape_list):
    print(f"--- Rendering {len(shape_list)} shapes ---")

    for shape in shape_list:
        shape.draw()

my_shapes = [Circle(), Triangle(), Square()]

render_screen(my_shapes)
```
