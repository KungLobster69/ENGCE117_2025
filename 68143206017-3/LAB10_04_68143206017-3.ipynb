```python
##WorkShop04
import tkinter as tk

root = tk.Tk()
root.title("Numpad")

display = tk.Entry(root, font=("Arial", 18))
display.grid(row=0, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

buttons = [
    '7', '8', '9',
    '4', '5', '6',
    '1', '2', '3'
]

row_val = 1
col_val = 0

for button_text in buttons:
    tk.Button(root, text=button_text, width=5, height=2).grid(row=row_val, column=col_val, padx=5, pady=5)
    col_val += 1
    if col_val > 2:
        col_val = 0
        row_val += 1

tk.Button(root, text='0', height=2).grid(row=4, column=0, columnspan=2, sticky="nsew", padx=5, pady=5)
tk.Button(root, text='.', width=5, height=2).grid(row=4, column=2, padx=5, pady=5)

root.mainloop()
```
