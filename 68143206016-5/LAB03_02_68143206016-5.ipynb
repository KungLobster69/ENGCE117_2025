```python
raw_names = [" somchai ", "alice", " BOB","c"]

output = list(filter(
    lambda name: len(name),
    map(lambda name: name.strip().lower(), raw_names)
))

print(output)
```

```python
cleaned_all = ['somchai', 'alice', 'bob', 'c']
filtered_names = list(filter(lambda name: len(name) > 2, cleaned_all))
print(filtered_names)
```

```python
raw_names = [" somchai ", "alice", " BOB"]

final_output = [name.strip().lower() 
                for name in raw_names 
                if len(name.strip()) > 1]

print(final_output)
```
