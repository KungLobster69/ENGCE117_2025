```python
def even_gen(limit):
    for num in range(limit + 1):
        if num % 2 == 0:
            yield num  


for n in even_gen(10):
    print(n)
```

```python
def frange(start, stop, step):

    current = start
    
    while current < stop:
        yield current    
        current += step   



for n in frange(0, 1, 0.5):
    print(n)

print("-" * 20)

for n in frange(0.5, 2.5, 0.5):
    print(n)
```

```python
import time

def number_generator(max_num):
    """
    Generator ที่ผลิตตัวเลขจาก 1 ถึง max_num
    """
    print("--- Gen1: เริ่มผลิตตัวเลข ---")
    for i in range(1, max_num + 1):
        print(f"Gen1: ผลิตเลข {i}")
        time.sleep(0.5)
        yield i
    print("--- Gen1: ผลิตเสร็จสิ้น ---")

def square_processor(input_generator):
    """
    Generator ที่รับข้อมูลจาก input_generator มายกกำลังสอง
    """
    print("--- Gen2: เริ่มประมวลผล ---")
    for number in input_generator:
        result = number ** 2
        print(f"Gen2: รับ {number} -> ยกกำลังสองได้ {result} -> ส่งต่อ")
        time.sleep(0.5)
        yield result
    print("--- Gen2: ประมวลผลเสร็จสิ้น ---")

gen1_output = number_generator(5) 

pipeline_output = square_processor(gen1_output)

print("\n=== เริ่มดึงข้อมูลจาก Pipeline (Gen2) ===\n")

for final_result in pipeline_output:
    print(f"*** Main Loop: รับผลลัพธ์สุดท้าย: {final_result} ***\n")
```
