```python
def sum_rec(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n + sum_rec(n-1)
    
sum_rec(5)
```

```python
def reverse_str(s):
    if len(s) == 0:
        return ""
    else:
        rest_of_string = s[-1]
        first_char = s[:-1]
    
    return rest_of_string + reverse_str(first_char)

reverse_str('ABC')
```

```python
def power(base: int, exp: int) -> int:
    
    if exp == 0:
        return 1
    return base * power(base, exp - 1)

base, exp = 0, 0

print(f'{base} ยกกำลัง {exp} = {power(base, exp)}')
```
