```python
import tkinter as tk

def submit():
    if accept_var.get():
        print(f"Username: {user_var.get()}")
    else:
        print("Please accept the terms before submitting!")

root = tk.Tk()
root.title("Registration Form")
root.geometry("300x200")


user_var = tk.StringVar()
accept_var = tk.BooleanVar()

tk.Label(root, text="Username:").pack(pady=(20, 0))
entry = tk.Entry(root, textvariable=user_var)
entry.pack(pady=5)

checkbox = tk.Checkbutton(root, text="I accept the terms", variable=accept_var)
checkbox.pack(pady=10)

btn_submit = tk.Button(root, text="Submit", command=submit)
btn_submit.pack(pady=10)

root.mainloop()
```
