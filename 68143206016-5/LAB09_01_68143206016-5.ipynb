```python
class RPGCharacter:
    def __init__(self,name, hp, mp):
        self.name = name
        self.hp = hp
        self.mp = mp

    def __str__(self):
        return f"Charactor = ({self.name}, {self.hp}, {self.mp})"

    def __add__ (self, other):
        if isinstance(other,RPGCharacter):
            New_name = f"{self.name} {other.name}"
            return RPGCharacter (New_name ,self.hp + other.hp, self.mp + other.mp)
        return NotImplemented
    

A = RPGCharacter("Hero",100,50)
E = RPGCharacter("Blind",100,50)
Heal = A.hp + 50
I = A + E


print(f"{A} + {E} = New {I}")
print(f"{A} + 50 Hp = {Heal} Hp")
```
