```python
#Lavel 1
inventory = {'apple':10, 'banana':20}

def check_price(item_name):
    return inventory.get(item_name, 0)

print(f'{check_price('durian')}')
```

```python
#Lavel 2
def add_stock(item, quantity):
    inventory[item] = inventory.get(item, 0) + quantity
    return inventory
x = add_stock('apple',5)
print(x)
```

```python
#Lavel 3
branch2 = {'banana':5, 'orange':10}
def merging(main, branch):
    for  item, quantity in branch.items():
        main[item] = main.get(item, 0) + quantity
        return main

print(merging(inventory, branch2))
```
