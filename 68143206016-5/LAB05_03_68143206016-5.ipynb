```python
class RobotPart:
    total_count = 0
    def __init__(self,name):
        self.name = name
        RobotPart.total_count += 1

category = "Mechanical"
r1 = RobotPart("Motor")
p1 = RobotPart("p1")


print(f"part : {r1.name} , Type : {category}")
print(f"Total parts in system : {RobotPart.total_count}")
print(f"Is it shared ?",p1.total_count is RobotPart.total_count)
```
