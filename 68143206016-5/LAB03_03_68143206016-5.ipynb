```python
nickname = "Pond"
content = (f"Nickname: {nickname}")
with open('note.txt', 'w', encoding='utf-8') as file:
    file.write(content)

print(f"เขียนข้อความ '{content}' ลงในไฟล์ note.txt เรียบร้อยแล้ว")

favorite_food = input("กรุณาป้อนอาหารที่คุณชอบ: ")

content_to_write = f"\nFavorite Food: {favorite_food}"

with open('note.txt', 'a', encoding='utf-8') as file:
    file.write(content_to_write)

print(f"บันทึกข้อมูลอาหารที่ชอบ ('{favorite_food}') ต่อท้ายไฟล์ note.txt เรียบร้อยแล้ว\n")

try:
    with open('note.txt', 'r', encoding='utf-8') as file:
        content = file.read()
    
    print("--- เนื้อหาในไฟล์ note.txt ---")
    print(content)
    print("------------------------------")

except FileNotFoundError:
    print("Error: ไม่พบไฟล์ note.txt กรุณาสร้างไฟล์และเขียนข้อมูลก่อน")
```
