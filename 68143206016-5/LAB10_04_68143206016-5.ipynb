```python
import tkinter as tk

root = tk.Tk()
root.title("Numpad Layout")

entry = tk.Entry(root, font=("Arial", 18), justify='right')
entry.grid(row=0, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

buttons = [
    '7', '8', '9',
    '4', '5', '6',
    '1', '2', '3'
]

row_val = 1
col_val = 0

for button_text in buttons:
    btn = tk.Button(root, text=button_text, width=5, height=2, font=("Arial", 12))
    btn.grid(row=row_val, column=col_val, padx=5, pady=5)
    
    col_val += 1
    if col_val > 2:
        col_val = 0
        row_val += 1

btn_zero = tk.Button(root, text="0", width=12, height=2, font=("Arial", 12))
btn_zero.grid(row=4, column=0, columnspan=2, padx=5, pady=5, sticky="nsew")

btn_dot = tk.Button(root, text=".", width=5, height=2, font=("Arial", 12))
btn_dot.grid(row=4, column=2, padx=5, pady=5)

root.mainloop()
```
