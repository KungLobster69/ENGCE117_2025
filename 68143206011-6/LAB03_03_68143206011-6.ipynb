```python
my_nickname = "Teerap"
with open('note.txt', 'w', encoding='utf-8') as f:
    f.write(f"Nickname: {my_nickname}\n")   
print("Level 1: File 'note.txt' created with nickname.")
```

```python

my_favorite_food = input("Enter your favorite food: ") 
with open('note.txt', 'a', encoding='utf-8') as f:
    f.write(f"Favorite Food: {my_favorite_food}\n") 
print("Level 2: Favorite food appended to 'note.txt'.")
```

```python
file_name = 'note.txt'

try:
    with open(file_name, 'r', encoding='utf-8') as f:
        content = f.read()
    
    print(f"\n--- Content of {file_name} ---")
    print(content, end="") 
    print("--------------------------------")
    
except FileNotFoundError:
    print(f"Error: File '{file_name}' not found.") 
    
print("Level 3: File content read and displayed.")
```
