```python
group_A = {'Reading', 'Coding', 'Gaming'}
group_B = {'Coding', 'Traveling', 'Gaming'}
common = group_A & group_B
print(f"Common: {common}")
```

```python
group_A = {'Reading', 'Coding', 'Gaming'}
group_B = {'Coding', 'Traveling', 'Gaming'}
only_A = group_A - group_B
symmetric_diff = group_A ^ group_B
print(f"Only A: {only_A}")
print(f"Diff: {symmetric_diff}")
```
