```python
def to_upper(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper
def add_brackets(func):
    def wrapper():
        result = func()        
        return f"[{result}]"  
    return wrapper
@add_brackets  
@to_upper    
def get_message():
    return "hello python"
print(get_message())
```
