```python
def even_gen(limit):
    for i in range(limit + 1):
        if i % 2 == 0:
            yield i

for num in even_gen(6):
    print(num)
```

```python
def frange(start, stop, step):
    current = float(start)
    while current < stop:
        yield current
        current += step
for val in frange(0, 1, 0.5):
    print(val)
```

```python
def producer():
    for i in range(1, 6):
        yield i
def square_processor(numbers):
    for n in numbers:
        yield n ** 2

pipeline = square_processor(producer())

for result in pipeline:
    print(result)
```
