```python
def sum_rec(n):
    if n == 1:
        return 1
    return n + sum_rec( n - 1)
n = 30
result = sum_rec (n)
print(f"(n) {result}")
```

```python
def reverse_str(s):
    if len(s) <= 1:
        return s
    return s[-1] + reverse_str(s[:-1])
print(reverse_str("123"))
```

```python
def power(base, exp):
    if exp == 0:
        return 1
    return base * power(base, exp - 1)
print(power(2, 3))
```
