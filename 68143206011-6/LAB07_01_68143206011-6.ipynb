```python
class Animal:
    def make_sound(self):
        print("... (Silence)")
```

```python
class Dog(Animal):
    def make_sound(self):
        print("Woof! Woof!")

class Cat(Animal):
    def make_sound(self):
        print("Meow~ Meow~")
```

```python
band = [Dog(), Cat(), Dog(), Animal()]

print("--- Zoo Chorus ---")
for animal in band:
    animal.make_sound()
```
