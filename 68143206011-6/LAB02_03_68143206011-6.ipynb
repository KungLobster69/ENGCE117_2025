```python
inventory = {'apple': 10, 'banana': 20 }
def check_stock(item_name):
    return inventory.get(item_name, 0)
print(f"Check Durian: {check_stock('durian')}")
```

```python
inventory = {'apple': 10, 'banana': 15}
def add_stock(item, quantity):
    current_stock = inventory.get(item, 0)
    inventory[item] = current_stock + quantity
add_stock('apple', 5)     
add_stock('banana', 5)   
add_stock('orange', 10)
add_stock('Mango', 20)
add_stock('Durian', 40)   
print(inventory)
```

```python
inventory = {'apple': 15, 'banana': 20, 'orange': 10}
branch2 = {'banana': 5, 'grape': 10, 'Magui': 20, } 
for item, quantity in branch2.items():
    current_stock = inventory.get(item, 0)
    inventory[item] = current_stock + quantity
print(inventory)
```
