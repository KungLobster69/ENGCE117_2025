```python
raw_names = [" somchai ", " alice ", " BOB", "c" , "Film HEhe"]
cleaned_map_result = list(map(lambda name: name.strip().lower(), raw_names))
print(f"Cleaned (Map): {cleaned_map_result}")

```

```python
names_to_filter = ["somchai", "alice", "bob", "c", "Film HEhe", "m", "mk",]
filtered_result = list(filter(lambda name: len(name) > 2, names_to_filter))
print(f"Filtered (Filter): {filtered_result}")
```

```python
raw_names = [" somchai ", "alice", " BOB", "c" ,"Film HEhe", "m", "mk","mk2", "mk3", ]
one_liner_result = [
    name.strip().lower()  
    for name in raw_names 
    if len(name.strip().lower()) > 2 
]
print(f"One-Liner (List Comp): {one_liner_result}")
```
