```python
def super_sum(*nums):
    return sum(nums)
result1 = super_sum(10, 20, 30)
result2 = super_sum(1, 2, 3, 4, 5, 6)
print (f"{result1} {result2}")

```

```python
def flexi_calc(mode, *args):
    if mode == "add":
        return sum(args)
    elif mode == "mult":
        result = 1
        for num in args:
            result *= num
        return result
    else:
        return "Error: Invalid mode."
result_add = flexi_calc("add", 5, 10, 15)
result_mult = flexi_calc("mult", 2, 3, 4)
print(f"Flexi Calc (add): {result_add}")
print(f"Flexi Calc (mult): {result_mult}")
```

```python
def stat_calc(mode, *args):
    if not args:
        return 0
    num_count = len(args)
    if mode == "avg":
        return sum(args) / num_count
    elif mode == "max":
        return max(args)
    else:
        return "Error: Invalid mode."
result_avg_full = stat_calc("avg", 10, 20, 30, 40, 50, 60,)
result_avg_empty = stat_calc("avg",)
result_max = stat_calc("max", 5, 12, 8, 20, 30, 40, 50, 60, 1000000000000000)
print(f"Stat Calc (avg, 10, 20, 30): {result_avg_full}")
print(f"Stat Calc (avg, empty): {result_avg_empty}")
print(f"Stat Calc (max, full): {result_max}")
```
