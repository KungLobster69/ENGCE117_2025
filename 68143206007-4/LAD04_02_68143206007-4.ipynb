```python
def even_gen(limit):
    for i in range(limit + 1):
        if i % 2 == 0:
            yield i

# ใช้งาน
for x in even_gen(10):
    print(x)
```

```python
def frange(start, stop, step):
    current = start
    while current < stop:
        yield current
        current += step

# ใช้งาน
for x in frange(0, 1, 0.5):
    print(x)
```

```python
def gen_numbers():
    for i in range(1, 6):
        yield i

def gen_square(source):
    for x in source:
        yield x ** 2

pipeline = gen_square(gen_numbers())

for value in pipeline:
    print(value)
```
