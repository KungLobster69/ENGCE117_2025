```python
food = input("อาหารที่ชอบ: ")

with open("note.txt", "a" , encoding='utf-8') as f:
    f.write("\nFavorite food: " + food)

print("Level 2: เขียนข้อมูลต่อท้ายสำเร็จ")
```

```python
# Level 3: Challenge — อ่านข้อมูลจากไฟล์แล้วแสดงผล

with open("note.txt", "r", encoding='utf-8') as f:
    data = f.read()

print("ข้อมูลในไฟล์ note.txt:\n")
print(data)
```
