```python
class Circle:
    def draw(self):
        print("วงกลม")

class Square:
    def draw(self):
        print("สี่เหลียม")

C = Circle()
S = Square()

C.draw()
S.draw()

```

```python
shapes = [Circle(), Square(), Circle(), Square()]
print("== เริ่มวาดรูปแบบ Loop ==")
for item in shapes:
    item.draw()
```

```python
class Triangle:
    def draw(self):
        print("สามเหลี่ยม")

def render_screen(shape_list):
    print (f"== Rendering {len(shape_list)} shapess ==")
    for shape in shape_list:
        shape.draw()
shape = [Circle(), Square(), Triangle(), ]
render_screen(shape)
        
    
```
