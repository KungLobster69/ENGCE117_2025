```python

class Robot:
    pass

r1 = Robot()
r1.name = "Commander"
r1.Battery = "100%"

r2 = Robot()
r2.name = "Commander"
r2.Battery = "100%"

# ตรวจสอบว่าเป็น object เดียวกันหรือไม่
if r1 is r2:
    print("Is r1 the same as r2? True")
else:
    print("Is r1 the same as r2? False")

print("Address r1:", id(r1))
print("Address r2:", id(r2))
```
