```python
class RobotPart:
    category = "Mechanical"
    total_count = 0   # Class Attribute

    def __init__(self, name):
        self.name = name
        RobotPart.total_count += 1

    def __str__(self):
        return f"Part: {self.name}, Type: {RobotPart.category}"


p1 = RobotPart("Motor")
p2 = RobotPart("Sensor")

print("Total parts in system:", RobotPart.total_count)

```

```python
p3 = RobotPart("Camera")

print("Is it shared?", p1.total_count is RobotPart.total_count)

```
