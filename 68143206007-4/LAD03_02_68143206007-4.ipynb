```python
raw_names = ["  somchai ", "alice", " BOB", "c"]
cleaner = list(map(lambda x: x.strip().lower(), raw_names))

print(cleaner)
```

```python
raw_names = ["  somchai ", "alice", " BOB", "c"]
f_names = list(filter(lambda x: len(x) > 1, cleaner))
print("Filtered names:", f_names)
```

```python
raw_names = ["  somchai ", "alice", " BOB", "c"]
cleaned_names = [x.strip().lower() for x in raw_names if len(x.strip()) > 1]
print("Filtered names:", cleaned_names)
```
