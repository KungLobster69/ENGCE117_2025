```python
def super_sum(*nums):
    return sum(nums)

super_sum(10 , 20 , 30)

```

```python

def flexi_calc(mode, *args):
    if mode == "add":
        return sum(args)
    elif mode == "mult":
        result = 1
        for n in args:
            result *= n
        return result
    else:
        return "Error: mode must be 'add' or 'mult'"

print(flexi_calc("add", 1, 2, 3))     # 6
print(flexi_calc("mult", 2, 3, 4))    # 24
```

```python

def stat_calc(mode, *args):
    if len(args) == 0:
        return 0

    if mode == "avg":
        return sum(args) / len(args)
    elif mode == "max":
        return max(args)
    else:
        return "Error: mode must be 'avg' or 'max'"

print(stat_calc("avg", 10, 20, 30))   # 20.0
print(stat_calc("avg"))               # 0
print(stat_calc("max", 5, 9, 3))      # 9
print(stat_calc("max"))               # 0
```
