```python
class RPGCharacter :
    def __init__(self,name,hp,mp):
        self.name, self.hp, self.mp = name, hp, mp
    
    def __str__(self):
        return f"{self.name}(HP:{self.hp}, MP:{self.mp})"
    
    def __add__(self, other):
        if isinstance(other, RPGCharacter):
            new_name = self.name + other.name
            new_hp = self.hp + other.hp
            new_mp = self.mp + other.mp
            return RPGCharacter(new_name, new_hp, new_mp)
        elif isinstance(other, int):
            new_hp = self.hp + other
            return RPGCharacter(self.name, new_hp, self.mp)
        else:
            raise TypeError("Unsupported operand type")


hero = RPGCharacter("Perseus", 100, 50)
villain = RPGCharacter("Medusa", 80, 20)

print(f"Original: {hero}")

fusion = hero + villain
print(f"Fusion:   {fusion}")

healed_hero = hero + 50
print(f"Healed:   {healed_hero}")
```
