```python
class Character:
    def __init__(self, name, hp, power):
        self.name = name
        self.hp = hp
        self.power = power

    def take_damage(self, amount):
        self.hp -= amount
        print(f"{self.name} took {amount} damage! Remaining HP: {self.hp}")

    def is_alive(self):
        return self.hp > 0



class Warrior(Character):
    def __init__(self, name, hp, power, defense):
        
        super().__init__(name, hp, power)
        self.defense = defense

    def take_damage(self, amount):
        
        reduced_damage = max(0, amount - self.defense)
        self.hp -= reduced_damage
        print(f"{self.name} (Warrior) blocked with {self.defense} def! Received {reduced_damage} dmg.")

class Mage(Character):
    def __init__(self, name, hp, power, mana):
        super().__init__(name, hp, power)
        self.mana = mana

    def attack(self):
        if self.mana >= 10:
            self.mana -= 10
            print(f"{self.name} casted Spell! (Power x2)")
            return self.power * 2
        return self.power


Batman = Warrior("Batman", hp=100, power=20, defense=5)
Superman = Mage("Superman", hp=80, power=15, mana=30)

print("--- The Battle Begins! ---")

while Batman.is_alive() and Superman.is_alive():
    print(f"\n>> {Superman.name}'s turn")
    dmg = Superman.attack()
    Batman.take_damage(dmg)
    
    if Batman.is_alive():
        print(f">> {Batman.name}'s turn")
        Superman.take_damage(Batman.power)

print("\n--- Game Over ---")

if Batman.is_alive():
    print(f"Winner: {Batman.name}")
else:
    print(f"Winner: {Superman.name}")
```

```python
class Character:
    def __init__(self, name, hp, power):
        self.name, self.hp, self.power = name, hp, power

    def take_damage(self, amount):
        self.hp -= amount
        print(f"{self.name} took {amount} dmg! HP: {self.hp}")

    @property
    def alive(self): return self.hp > 0

class Warrior(Character):
    def __init__(self, name, hp, power, defense):
        super().__init__(name, hp, power)
        self.defense = defense

    def take_damage(self, amount):
        print(f"{self.name} blocked {self.defense} dmg.")
        super().take_damage(max(0, amount - self.defense))

class Mage(Character):
    def __init__(self, name, hp, power, mana):
        super().__init__(name, hp, power)
        self.mana = mana

    def attack(self):
        if self.mana >= 10:
            self.mana -= 10
            print(f"{self.name} cast Spell! (x2 Dmg)")
            return self.power * 2
        return self.power

# --- Main Game ---
bat, sup = Warrior("Batman", 100, 50, 5), Mage("Superman", 80, 15, 30)

print("--- Fight! ---")
while bat.alive and sup.alive:
    # Superman Turn
    bat.take_damage(sup.attack())
    if not bat.alive: break
    
    # Batman Turn
    sup.take_damage(bat.power)

print(f"\nWinner: {bat.name if bat.alive else sup.name}")
```
