```python
##Level 3 (Hard)

class RobotPart:
    category = "Mechanical"
    total_count = 0

    def __init__(self, name):
        self.name = name
        RobotPart.total_count += 1

p1 = RobotPart("Motor")
p2 = RobotPart("Gear")
p3 = RobotPart("Metal")

is_shared = (p3.total_count is RobotPart.total_count)

print(f'Is it shared?: {is_shared}')
```
