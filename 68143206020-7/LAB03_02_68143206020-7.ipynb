```python
raw_names = [" somchai " , "alice" , " BOB" , "c"]

Cleaned = list(map(lambda name: name.strip().lower(), raw_names))   

print(Cleaned)
```

```python
raw_names = [" somchai " , "alice" , " BOB" , "c"]

Filter = list(filter(lambda name: len(name) > 2, raw_names))

print(Filter)
```

```python
raw_names = [" somchai " , "alice" , " BOB" , "c"]

last_name =  [name.strip().lower() for name in raw_names if len(name.strip()) > 2]

print(last_name)
```
