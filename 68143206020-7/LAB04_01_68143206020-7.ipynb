```python
#Level 1

def sum_rec(n):
    
    if n == 1:
        return 1
    else:
        return n + sum_rec(n - 1)

n = 5
result = sum_rec(n)
print(f"ผลรวมตั้งแต่ 1 ถึง {n} คือ: {result}")
```

```python
#Level 2

def reverse_str(s):
    if len(s) == 0:
        return ""
    else:
        return s[-1] + reverse_str(s[:-1])

text = "123"
result = reverse_str(text)
print(f"กลับด้าน '{text}' เป็น -> '{result}'")
```

```python
#Level 3

def power(base, exp):
    if exp == 0:
        return 1
    else:
        return base * power(base, exp - 1)

b = 4
e = 3
result = power(b, e)
print(f"{b} ยกกำลัง {e} คือ: {result}")
```
