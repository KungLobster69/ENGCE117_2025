```python
#Level 1
from collections import Counter

text = "apple banana apple orange banana apple"
words = text.split()
word_counts = Counter(words)

most_common_word = word_counts.most_common(1)

print(f"Result:{most_common_word}")
#Level 2
from collections import defaultdict

names = ["Alice", "Bob", "Charlie", "David", "Eve"]

names = defaultdict(list)
names['Alice','David'] = 5
names['Bob', 'Eve'] = 3
names['Charlie'] = 7

inverted = {v:k for k,v in names.items()}

print(f"Output:{inverted}")
#Level 3
from collections import deque

log_system = deque(maxlen=3)

inputs = ['Log1', 'Log2', 'Log3', 'Log4', 'Log5']

for item in inputs:
    log_system.append(item)
    print(f"เพิ่ม {item}: {list(log_system)}")

print("\n--- ผลลัพธ์ (Output) ---")
print(list(log_system))
```
