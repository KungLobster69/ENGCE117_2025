```python
class Animal:
    def make_sound(self):
        print ("... (Silence)")


class Dog(Animal):
    def make_sound(self):
        print ("Woof! Woof!")

class Cat(Animal):
    def make_sound(self):
        print ("Meow~ Meow~")


band = [Dog(), Cat(), Dog(), Animal()]

for animal in band:
    animal.make_sound()
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager(Employee):
    def __init__(self, name, salary, dept):
        # ต้องส่ง name, salary ไปให้แม่สร้าง
        super().__init__(name, salary)
        self.dept = dept  # ส่วนขยายของลูก

    def show_info(self):
        print(f"Manager: {self.name}, Dept: {self.dept}")

m = Manager("Alice", 50000, "IT")
m.show_info() 
# ถ้าลืม super().__init__ จะ Error ว่าไม่มี attribute name/salary [cite: 4163-4172]
```
