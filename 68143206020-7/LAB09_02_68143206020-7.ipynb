```python
class Playlist:
    def __init__(self, songs):
        self.songs = list(songs)

    def __len__(self):
        return len(self.songs)

    def __getitem__(self, index):
        if isinstance(index, slice):
            sliced_songs = self.songs[index]
            return Playlist(sliced_songs)
        
        return self.songs[index]

    def __setitem__(self, index, value):
        self.songs[index] = value

    def __contains__(self, song):
        return song in self.songs

    def __str__(self):
        return f"Playlist({self.songs})"


my_playlist = Playlist(["Song A", "Song B", "Song C", "Song D"])
print("Original:", my_playlist)

print(f"จำนวนเพลง: {len(my_playlist)}")  
print(f"เพลงแรก: {my_playlist[0]}")      

my_playlist[0] = "New Song"             
print("After Edit:", my_playlist)
print(f"Is 'Song B' in playlist?: {'Song B' in my_playlist}") 

sub_playlist = my_playlist[1:3]          

print(f"Sub Playlist: {sub_playlist}")
print(f"Type of Sub Playlist: {type(sub_playlist)}")
```
