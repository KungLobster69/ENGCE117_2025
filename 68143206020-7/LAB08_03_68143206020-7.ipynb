```python
from abc import ABC, abstractmethod

class IoTDevice(ABC):
    @abstractmethod
    def turn_on(self):
        pass
    @abstractmethod
    def turn_off(self):
        pass

class SmartLight(IoTDevice):
    def turn_on(self):
        print("💡 ไฟเปิด")
    def turn_off(self):
        print("🌑 ไฟปิด")

obj = SmartLight()
obj.turn_on()
obj.turn_off()

class SmartSpeaker(IoTDevice):
    def turn_on(self):
        try:
            obj = SmartSpeaker() 
            obj.turn_on()
        except TypeError:
            print("❌ Error: Can't instantiate abstract class SmartSpeaker with abstract method turn_off")


def test_device(device):
    if isinstance(device, IoTDevice):
        device.turn_on()
        device.turn_off()
    else:
        print("⚠️ ไม่ใช่อุปกรณ์ IoT ที่รองรับ")

print("--Testing SmartLight--")
light = SmartLight()
test_device(light)
```

```python
from abc import ABC, abstractmethod

class Shape(ABC): # 1. สืบทอดจาก ABC
    @abstractmethod # 2. แปะป้ายว่าเป็น Abstract Method
    def area(self):
        pass

class Square(Shape):
    def __init__(self, side):
        self.side = side
    
    def area(self): # 3. ลูกต้องเขียนทับ (Implement)
        return self.side * self.side

# s = Shape()      # Error! สร้างแม่ไม่ได้
sq = Square(4)     # OK
print(sq.area())   # Output: 16 [cite: 4682-4688]
```
