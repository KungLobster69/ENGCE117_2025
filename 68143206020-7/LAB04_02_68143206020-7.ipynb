```python
#Level 1

def even_gen(limit):
    for i in range(1, limit + 1):
        if i % 2 == 0:
            yield i  

for num in even_gen(20):
    print(num, end=" ")  
print("\n")
```

```python
#Level 2

def frange(start, stop, step):
    current = start
    while current < stop:
        yield current     
        current += step   


gen = frange(0, 5, 1)
print(list(gen))
```

```python
#Level 3

def number_producer():
    for i in range(1, 6):
        yield i

def square_processor(input_generator):
    for num in input_generator:
        yield num ** 2

source = number_producer()       
pipeline = square_processor(source) 

for result in pipeline:
    print(f"Factory Output: {result}")
```
