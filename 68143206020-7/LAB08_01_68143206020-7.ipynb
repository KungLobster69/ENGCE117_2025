```python
class Circle:
    def draw(self):
        print(" 🔴 วาดวงกลม ")

class Square:
    def draw(self):
        print(" 🟦 วาดสี่เสี่หลี่ยม ")

C = Circle()
S = Square()

C.draw()
S.draw()

shapes = [Circle(), Square(), Circle(), Square()]
for item in shapes:
    item.draw()

class Triangle:
    def draw(self):
        print("🔺 วาดสามเหลี่ยม")

def render_screen(shape_list):
    for shape in shape_list:
        shape.draw()

shapes = [Circle(), Square(), Triangle()]
for item in shapes:
    item.draw()
```
