```python
def super_sum(*nums):
    return sum(nums)
print(super_sum(4, 6, 8, 10, 12))         
print(super_sum( 20, 40, 60, 80)) 
print(super_sum())
```

```python
def flexi_calc(mode,*args):
 if mode == "add":
        return sum(args)
 elif mode == "mult":
        result = 1 
        for number in args:
            result *= number 
        return result       
 else :
    print("กรอกให้ถูก")

flexi_calc("mult",10,20,30)
```

```python
def stat_calc(mode, *args):
    if  args == None:
        return 0
    if mode == "avg":
        return sum(args) / len(args)
    elif mode == "max":
        return max(args)
   
    

print(stat_calc("avg", 10, 20, 30)) 
print(stat_calc("avg",0))
print(stat_calc("max",1,2,3)) 
```
