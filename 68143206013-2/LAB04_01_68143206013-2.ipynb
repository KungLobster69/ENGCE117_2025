```python
def sum_rec(n):
    if n == 0:
        return 0
    return n + sum_rec(n-1)   
sum_rec(5 )        
```

```python
### ข้อ2 ไม่loop
def reverse_str(s):
   return s[::-1]
reverse_str("12345")
```

```python
### ข้อ2 loop
def reverse_str(s):
    if len(s) == 0:
        return ''
    return s[-1] + reverse_str(s[:-1])
reverse_str("12345")
```
