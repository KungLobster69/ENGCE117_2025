```python
import tkinter as tk
def change_text():
    label.config(text="คลิกแล้วจ้า!", fg="red")

root = tk.Tk()
root.title("My First App")            
root.geometry("200x150")  
btn = tk.Button(root, text="ยืนยัน",command = change_text)
btn.pack(pady=10)
label = tk.Label(root, text="ยังไม่ได้คลิก")
label.pack(pady=10)
root.mainloop()  

```
