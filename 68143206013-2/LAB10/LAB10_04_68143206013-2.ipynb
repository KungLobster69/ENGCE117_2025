```python
import tkinter as tk

root = tk.Tk()
root.title("Numpad")

entry = tk.Entry(root, width=25, font=("Arial", 16))
entry.grid(row=0, column=0, columnspan=3, padx=10, pady=10)

def add(n):
    entry.insert(tk.END, n)

buttons = [
    ("7", 1, 0), ("8", 1, 1), ("9", 1, 2),
    ("4", 2, 0), ("5", 2, 1), ("6", 2, 2),
    ("1", 3, 0), ("2", 3, 1), ("3", 3, 2),
]

for (text, r, c) in buttons:
    tk.Button(root, text=text, font=("Arial", 16), width=5, height=2,
              command=lambda t=text: add(t)).grid(row=r, column=c, padx=5, pady=5)

tk.Button(root, text="0", font=("Arial", 16), width=13, height=2,
          command=lambda: add("0")).grid(row=4, column=0, columnspan=2, padx=5, pady=5)

tk.Button(root, text=".", font=("Arial", 16), width=5, height=2,
          command=lambda: add(".")).grid(row=4, column=2, padx=5, pady=5)

root.mainloop()
```
