```python
import tkinter as tk
def show_text():
    if tkbl.get():  
        username = tkstr.get()
        finis.config(text=f"ยินดีต้อนรับ {username}")
    else:
        finis.config(text="กรุณายอมรับเงื่อนไขก่อน!")

root = tk.Tk()
root.geometry("300x200")
root.title("Registration Form")
tkstr = tk.StringVar()
tkbl = tk.BooleanVar()
label = tk.Label(root,text = "Username:")
label.pack()
entry = tk.Entry(root,show = "",textvariable=tkstr)
entry.pack(pady = 5)
checkbutton = tk.Checkbutton(root,text = "I accept the terms",variable = tkbl)
checkbutton.pack()
btn = tk.Button(root, text = "Submit",command = show_text,)
btn.pack(pady = 10)
finis = tk.Label(root,text = " ")
finis.pack()
root.mainloop()
```
