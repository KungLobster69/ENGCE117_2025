```python
def super_sum(*num):
    return sum(num)
super_sum(10,10,6)
```

```python
def flexi_calc(mode,*args):
    if mode == '+':
        a = sum(args)
    elif mode == '*':
        a = 1
        for i in args:   
            a *= i 
    else:
        print("error")
    return a
flexi_calc('*',2,9)
```

```python
def stat_calc(mode,*args): 
        if len(args) == 0:
            return 0
        if mode == 'avg':
            avg = sum(args)/len(args)
            return avg,max(args)
        else :
            return 0
stat_calc('avg')
```
