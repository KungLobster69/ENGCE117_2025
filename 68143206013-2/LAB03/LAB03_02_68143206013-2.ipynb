```python
raw_name = [" somchai ", "alice"," BOB","c"]
new_rawname = list(map(lambda name: name.strip().lower(),raw_name ))
print(new_rawname)
```

```python
raw_name = [" somchai ", "alice"," BOB","c"]
new_rawname = list(filter(lambda x: len(x) >= 2 ,raw_name))
print(new_rawname)
```

```python
raw_name = [" somchai ", "alice", " BOB", "c"]
new_rawname = list(map(lambda n: n.strip().lower(),filter(lambda n: len(n.strip()) >= 2, raw_name)))
print(new_rawname)
```
