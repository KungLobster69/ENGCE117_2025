```python
class Animal():
    def make_sound(self):
        print("...slient")
class Dog(Animal):
    def make_sound(self):
        print("woof woof")
class Cat(Animal):
    def make_sound(self):
        print("meow meow") 
band = [Dog(),Cat(),Dog(),Animal()]  
for i in band:
    i.make_sound()      
```
