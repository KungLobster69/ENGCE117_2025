```python
class Character:
    def __init__(self, name, hp, power):
        self.name = name
        self.hp = hp
        self.power = power

    def take_damage(self, amount):
        if amount < 0:
            amount = 0
        self.hp -= amount
        if self.hp < 0:
            self.hp = 0
        print(f"{self.name} ได้รับความเสียหาย {amount} เหลือ HP = {self.hp}")

    def is_alive(self):
        return self.hp > 0

    def attack(self, other):
        if not self.is_alive():
            print(f"{self.name} โจมตีไม่ได้ เพราะล้มแล้ว")
            return
        print(f"{self.name} โจมตี {other.name} ด้วยพลัง {self.power}")
        other.take_damage(self.power)
class Warrior(Character):
    def __init__(self, name, hp, power, defense):
        super().__init__(name, hp, power)
        self.defense = defense

    def take_damage(self, amount):
        if amount < 0:
            amount = 0
        reduced = amount - self.defense
        if reduced < 0:
            reduced = 0
        self.hp -= reduced
        if self.hp < 0:
            self.hp = 0
        print(
            f"{self.name} ป้องกันด้วย DEF {self.defense} "
            f"ความเสียหายจริง = {reduced} เหลือ HP = {self.hp}"
        )
class Mage(Character):
    def __init__(self, name, hp, power, mana, mana_per_cast=5, bonus_damage=10):
        super().__init__(name, hp, power)
        self.mana = mana
        self.mana_per_cast = mana_per_cast
        self.bonus_damage = bonus_damage

    def attack(self, other):
        if not self.is_alive():
            print(f"{self.name} โจมตีไม่ได้ เพราะล้มแล้ว")
            return

        if self.mana >= self.mana_per_cast:
            self.mana -= self.mana_per_cast
            dmg = self.power + self.bonus_damage
            print(
                f"{self.name} ร่ายเวท! ใช้มานา {self.mana_per_cast} (เหลือ {self.mana}) "
                f"สร้างความเสียหาย {dmg} ให้ {other.name}"
            )
            other.take_damage(dmg)
        else:
            print(f"{self.name} มานาไม่พอสำหรับเวท โจมตีปกติด้วยพลัง {self.power}")
            other.take_damage(self.power)

def battle(p1: Character, p2: Character, first="p1"):
    print(f"เริ่มการต่อสู้: {p1.name} (HP={p1.hp})  VS  {p2.name} (HP={p2.hp})")
    attacker, defender = (p1, p2) if first == "p1" else (p2, p1)
    turn = 1

    while p1.is_alive() and p2.is_alive():
        print(f"\n เทิร์น {turn}: {attacker.name} เป็นฝ่ายโจมตี ")
        attacker.attack(defender)
        if not defender.is_alive():
            break
        attacker, defender = defender, attacker
        turn += 1
    winner = p1 if p1.is_alive() else p2
    print(f"จบการต่อสู้! ผู้ชนะคือ {winner.name} (HP เหลือ {winner.hp})")
    return winner
if __name__ == "__main__":
    w = Warrior(name="BEST", hp=40, power=6, defense=4)
    m = Mage(name="Bay", hp=25, power=5, mana=12, mana_per_cast=5, bonus_damage=8)
    battle(w, m, first="p1")
```
