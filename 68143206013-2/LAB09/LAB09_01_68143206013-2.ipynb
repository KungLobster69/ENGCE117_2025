```python
###LEVEL3-2
class RPGCharacter:
    def __init__(self,name,hp,mp):
        self.name = name
        self.hp = hp
        self.mp = mp
    def __str__(self):
        return f"{self.name}(HP:{self.hp},MP:{self.mp})"
    def __add__(self,x):
        if isinstance(x,RPGCharacter):
            return RPGCharacter(self.name + x.name , self.hp + x.hp , self.mp + x.mp)
        elif isinstance(x,int):
            healhp = self.hp + x
            return healhp
char1 = RPGCharacter("bes",100,50)                
char2 = RPGCharacter("bor",200,100)
char3 = char2 + char1
print(f"{char1}+{char2} = {char3}")
healchar1 = char3 + 50000
print(f"{char3}heal = HP:{healchar1}")
```
