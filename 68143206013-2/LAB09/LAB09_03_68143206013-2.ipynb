```python
def to_upper(x):
    def wrapper(*args, **kwargs):
        result = x(*args, **kwargs)
        return f"{str(result).upper()}"
    return wrapper

@to_upper
def get_string(y):
    return y.upper()

get_string("sadadasa")
```

```python
def add_brackets(x):
    def wrapper(*args, **kwargs):
        result = x(*args, **kwargs)
        return f"[{str(result).upper()}]"
    return wrapper

@add_brackets
def get_string(y):
    return y

get_string("sasasasasa")
```
