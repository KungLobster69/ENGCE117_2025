```python
class square():
    def __init__(self,side):
        self.__side = side

    @property
    def side(self):
        return self.__side
    
    @side.setter
    def side(self,limitside):
        if limitside > 0 :
            self.__side = limitside
        else:
            raise ValueError("error")
    @property
    def calside(self,):
        return self.__side * self.__side 
s = square(4)
print(f"side = {s._square__side} , area = {s.calside}") 
s.side = 5
print(f"New Area = {s.calside}")
s.side = -10
print(f"New Area = {s.calside}")





       
```
