```python
inventory = {'apple':10 , 'banana': 20}
def check_price(item_name):
    return  inventory.get(item_name , 0) 
print(check_price('Durian'))

```

```python
inventory = {'apple':10 , 'banana': 20}
def add_stock(item,quantity):
        inventory[item] = inventory.get(item, 0) + quantity    
add_stock('apple', 5)
print(inventory)       
```

```python
inventory = {'apple':10 , 'banana': 20}
branch2 = {'banana':5 , 'orange':10}
for item,quantity in branch2.items():
    if item in inventory:
        inventory[item] += quantity
    else:
        inventory[item] = quantity
print(inventory)


    

     
```
