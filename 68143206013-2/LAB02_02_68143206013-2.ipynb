```python
numbers = (1,2,3,4,5)
def get_stats(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    return min_val,max_val
total1,total2 = get_stats(numbers)
print(f"Min: {total1} Max: {total2}")
```
