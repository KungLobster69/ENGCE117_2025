```python
def even_gen(limit):
    for x in range(limit):
        if x%2 == 0:
            yield x 
for num in even_gen(7):
    print(num)

            
```

```python
def frange(start, stop, step):
    while start < stop:
        yield round(start, 10)  
        start += step
for num in frange(0, 1, 0.5):
    print(num)

```

```python
def pipeline():
    for i in range(1,5):
        yield i
p = pipeline()
def num_xx(p):
    for num in p:
        y = num**2
        yield y
for a in num_xx(p):
    print(a)

        
```
