```python
class robot():
    def __init__ (self,name,model,battery):
        self.name = name
        self.model = model
        self.battery = battery
r1 = robot("R-","03",15)
print(f"Robot:{r1.name}{r1.model} Battery:{r1.battery}%") 
if r1.battery < 20:
    print(f"Warning:{r1.name}{r1.model} has Low Battery!")         
```
