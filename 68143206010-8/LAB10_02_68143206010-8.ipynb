```python
import tkinter as tk

root = tk.Tk()
root.title("Workshop 2")
root.geometry("200x150")

def change_text():
    label.config(text="คลิกแล้วจ้า!", fg="red")

label = tk.Label(root, text="ยังไม่ได้คลิก")
label.pack(pady=10)

btn = tk.Button(root, text="คลิกฉันสิ!", command=change_text)
btn.pack()

root.mainloop()
```
