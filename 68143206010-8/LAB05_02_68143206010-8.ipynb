```python
class Robot:
    def __init__(self, name, moduel):
        self.name = name
        self.module = moduel

r1 = Robot("R-01" , "Explorer")
print(f"Robot: {r1.name} , {r1.module}")
```

```python
class Robot:
    def __init__(self, name, moduel, battery=100):
        self.name = name
        self.module = moduel
        self.battery = battery

r2 = Robot("R-02" , "battery ")
print(f"Robot: {r2.name} ,battery {r2.battery}")
```

```python
class Robot:
    def __init__(self, name, module, battery):
        self.name = name
        self.module = module
        self.battery = battery

        if self.battery < 20:
            print(f"Warning: {self.name} has Low Battery!")

r3 = Robot("R-03", "Power Module", 19)
```
