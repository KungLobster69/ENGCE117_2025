```python
class RobotPart:
    category = "Mechanical"   # Class Attribute (ใช้ร่วมกันทุก object)

    def __init__(self, name):
        self.name = name      # Instance Attribute (ของใครของมัน)

    def show_info(self):
        print(f"Part: {self.name}, Type: {RobotPart.category}")

part1 = RobotPart("Motor")
part1.show_info()
```

```python
class RobotPart:
    category = "Mechanical"
    total_count = 0

    def __init__(self, name):
        self.name = name
        RobotPart.total_count += 1
part1 = RobotPart("Motor")
part2 = RobotPart("Gear")

print("Total parts in system:", RobotPart.total_count)
```

```python
# สร้างชิ้นส่วนเพิ่มอีก 1 ชิ้น
p3 = RobotPart("Camera")

# ตรวจสอบว่าค่า total_count ของ object และของคลาสเป็นค่าเดียวกันหรือไม่
print("Is it shared?", p3.total_count is RobotPart.total_count)
```
