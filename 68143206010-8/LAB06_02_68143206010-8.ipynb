```python
class Square:
    def __init__(self, side):
        self.__side = side 

    @property
    def side(self):
        return self.__side

    @side.setter
    def side(self, new_side):
        if new_side > 0:
            self.__side = new_side
        else:
            raise ValueError("ความยาวด้านต้องมากกว่า 0")
    @property
    def area(self):
        return self.__side * self.__side
    
    s = Square(4)
    print(f"Side: {s.side}, Area: {s.area}") # 4, 16

    s.side = 5
    print(f"New Area: {s.area}") # 25 (เปลี่ยนตามอัตโนมัติมัติ)

    # s.side = -10 # Error: ความยาวด้าด้นต้องมากกว่าว่ 0
    # s.area = 100 # Error: AttributeError (Read-only)
```
