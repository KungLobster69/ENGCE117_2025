```python
currencies = {'USD': 35, 'EUR': 38, 'JPY': 0.25}
inverted = {value: key for key, value in currencies.items()}
filtered_inverted = {k: v for k, v in inverted.items() if k in (35, 38)}
print("Inverted:", filtered_inverted)
```
