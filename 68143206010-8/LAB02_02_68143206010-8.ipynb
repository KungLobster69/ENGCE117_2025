```python
def get_stats(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    return (min_val, max_val)

nums = [1, 2, 3, 4, 5]
min_val, max_val = get_stats(nums)

print("Min:", min_val, ", Max:", max_val)
```

```python
def get_stats(numbers):
    low = min(numbers)
    high = max(numbers)

    others = [n for n in numbers if n != low and n != high]

    return low, high, others

nums = [10, 20, 30, 40, 50, 60]
low, high, others = get_stats(nums)

print(f"Low: {low}, High: {high}, Others: {others}")
```
