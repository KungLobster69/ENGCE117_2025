```python
def sum_rec(n):
    if n <= 0:
        return 0
    else:
        return n + sum_rec(n - 1)
print(sum_rec(5)) 
```

```python
def reverse_str(s):
    if len(s) == 0:
        return ""

    else:
        last_char = s[-1]
        remaining_str = s[:-1]
        return last_char + reverse_str(remaining_str)
print(reverse_str("ABC")) 
```

```python
def power(base, exp):
    if exp == 0:
        return 1
    else:
        return base * power(base, exp - 1)
print(power(5, 3))   
```
