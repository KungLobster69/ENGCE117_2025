```python
def even_gen(limit):
    n = 0
    while n < limit:
        if n % 2 == 0:
            yield n
        n += 1
for even_number in even_gen(11):
    print(even_number)
```

```python
def frange(start, stop, step):
    current = start
    if step > 0:
        while current < stop:
            yield current
            current += step
    elif step < 0:
        while current > stop:
            yield current
            current += step

for f_num in frange(0, 1, 0.5):
    print(f"{f_num:.1f}") 
```

```python
def pipeline():
    for i in range(1,5):
        yield i
p = pipeline()
def num_xx(p):
    for num in p:
        y = num**2
        yield y
for a in num_xx(p):
    print(a)
```
