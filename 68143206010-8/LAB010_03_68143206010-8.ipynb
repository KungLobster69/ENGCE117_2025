```python
import tkinter as tk

root = tk.Tk()
root.title("Registration Form")
root.geometry("300x200")


username_var = tk.StringVar()
accept_var = tk.BooleanVar()

def submit():
    if accept_var.get():
        print("Username:", username_var.get())
    else:
        print("กรุณายอมรับเงื่อนไขก่อน!")

tk.Label(root, text="Username:").pack(pady=5)
tk.Entry(root, textvariable=username_var).pack()                             
tk.Checkbutton(root, text="I accept the terms",
               variable=accept_var).pack(pady=5)

tk.Button(root, text="Submit", command=submit).pack(pady=10)

root.mainloop()
```
