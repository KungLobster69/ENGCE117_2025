```python
def super_sum (*nums):
    return sum(nums)
print(super_sum (10, 20, 30))
print(super_sum())
   

```

```python
def flexi_calc(mode, *args):
    if mode == "add":
        total = 0
        for n in args:
            total += n
        return total
    
    elif mode == "mult":
        result = 1
        for n in args:
            result *= n
        return result
    
    else:
        return "ERROR: unknown mode"

print(flexi_calc("add", 1, 2, 3))     # 6
print(flexi_calc("mult", 2, 3, 4))    # 24
print(flexi_calc("xxx", 1, 2, 3))     # ERROR
```

```python
def stat_calc(mode,*args):
    # ถ้า args ว่าง ให้คืนค่า 0
    if len(args) == 0:
        return 0

    if mode == "avg":
        return sum(args) / len(args)

    elif mode == "max":
        return max(args)

    else:
        return "ERROR: unknown mode"
print(stat_calc("avg", 10, 20, 30 ))  # คาดหวัง 20.0
print(stat_calc("avg"))              # คาดหวัง 0


```
