```python
import tkinter as tk

root = tk.Tk()
root.title("Numpad")

entry = tk.Entry(root, justify="right")
entry.grid(row=0, column=0, columnspan=3, padx=5, pady=5, sticky="ew")

numbers = [
    ('1', 1, 0), ('2', 1, 1), ('3', 1, 2),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2),
    ('7', 3, 0), ('8', 3, 1), ('9', 3, 2),
]

for (text, row, col) in numbers:
    tk.Button(root, text=text, width=5).grid(
        row=row, column=col, padx=5, pady=5
    )

tk.Button(root, text='0', width=12).grid(
    row=4, column=0, columnspan=2, padx=5, pady=5
)

tk.Button(root, text='.', width=5).grid(
    row=4, column=2, padx=5, pady=5
)

root.mainloop()
```
