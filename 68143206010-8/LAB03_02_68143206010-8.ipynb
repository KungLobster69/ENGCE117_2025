```python
raw_names = [" somchai ", "alice", " BOB", "c"]
cleaned = list(map(lambda x: x.strip().lower(), raw_names))
filtered = list(filter(lambda x: len(x) > 2, cleaned))

print(filtered)
```

```python
raw_names = [" somchai ", "alice", " BOB", "c"]
result = [name.strip().lower() for name in raw_names if len(name.strip()) > 2]
print(result)

```
