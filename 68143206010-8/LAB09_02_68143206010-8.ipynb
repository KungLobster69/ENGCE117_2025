```python
class Playlist:
    def __init__(self, songs):
        self.songs = list(songs)
        
    def __len__(self):
        return len(self.songs)

    def __getitem__(self, i):
        return Playlist(self.songs[i]) if isinstance(i, slice) else self.songs[i]

    def __setitem__(self, i, value):
        self.songs[i] = value

    def __contains__(self, item):
        return item in self.songs

    def __str__(self):
        return str(self.songs)
    
   
p = Playlist(["Song A", "Song B", "Song C"])

print(len(p))        
print(p[0])          
p[0] = "New Song"
print(p)             
print("Song B" in p) 
print(p[0:2])     
```
