```python
class Animal:
    def make_sound(self):
        print("... (Silence)")

class Dog (Animal):
    def make_sound(self):
        print("Woof! Woof!")  

class Cat (Animal):
    def make_sound(self):
        print("Meow!")

zoo = [Dog(), Cat(), Dog(), Animal()]       

print("--- Zoo ChorUS ---")
for animal in zoo:
    animal.make_sound()
```
