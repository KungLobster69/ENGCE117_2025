```python
class RPGCharacter:
   
    def __init__(self, name, hp, mp):
        self.name = name
        self.hp = hp
        self.mp = mp

    def __str__(self):
        return f"{self.name} (HP: {self.hp}, MP: {self.mp})"

    def __add__(self, other):
       
        if isinstance(other, int): 
            return RPGCharacter(self.name, self.hp + other, self.mp)
        elif isinstance(other, RPGCharacter):
        
            new_name = self.name + other.name
            new_hp = self.hp + other.hp
            new_mp = self.mp + other.mp
            return RPGCharacter(new_name, new_hp, new_mp)
        
        return NotImplemented

c1 = RPGCharacter("Hero", 100, 50)
c2 = RPGCharacter("Mage", 80, 120)

print(f"L1: {c1}")            
print(f"L2: {c1 + c2}")       
print(f"L3: {c1 + 50}")        
        
```
