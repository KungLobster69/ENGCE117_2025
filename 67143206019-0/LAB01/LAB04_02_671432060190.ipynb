```python
def even_gen(limit):
    for num in range(limit + 1):
        if num % 2 == 0 :
            yield num
my_evens = even_gen(10)
for val in my_evens:
    print(val)
```

```python
def frange(start, stop, step):
    current = start
    while current < stop:
        yield round(current, 10)
        current += step
for value in frange(0, 1, 0.5):
    print(value)
```

```python
def gen1_producer():
    """Gen1: Produces numbers 1-5"""
    for i in range(1, 6):
        print(f"Gen1 producing: {i}")
        yield i

def gen2_processor(numbers):
    """Gen2: Takes numbers -> squares them -> passes them on"""
    for n in numbers:
        squared = n ** 2
        print(f"  Gen2 squaring {n} -> {squared}")
        yield squared


pipeline = gen2_processor(gen1_producer())


print("--- Starting Pipeline ---")
for final_value in pipeline:
    print(f"    Final Output: {final_value}")
```
