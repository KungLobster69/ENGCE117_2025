```python
def get_stats(number):
    minl = min (number)
    maxl = max (number)
    return minl,  maxl
s = [1, 2, 3, 4, 5]
minl, maxl = get_stats(s)
print(f"Min: {minl}, Max: {maxl}")
    
```
