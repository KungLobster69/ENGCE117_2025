```python
import tkinter as tk

root = tk.Tk()
root.title("Workshop 2")
root.geometry("400x300")


label = tk.Label(root, text="ยังไม่ได้คลิก", font=("Arial", 14))
label.pack(pady=10)


button = tk.Button(root, text="คลิกฉันสิ!")
button.pack(pady=10)

root.mainloop()
```

```python
import tkinter as tk


def change_text():
    label.config(text="คลิกแล้วจ้า!")

root = tk.Tk()
root.title("Workshop 2")
root.geometry("400x300")

button = tk.Button(root, text="คลิกฉันสิ!", command=change_text)
button.pack(pady=10)


label = tk.Label(root, text="ยังไม่ได้คลิก", font=("Arial", 14))
label.pack(pady=10)


root.mainloop()
```

```python
import tkinter as tk

def change_text():
    label.config(text="คลิกแล้วจ้า!", fg="red")

root = tk.Tk()
root.title("Workshop 2: Advanced")

root.geometry("200x150")


button = tk.Button(root, text="คลิกฉันสิ!", command=change_text)
button.pack(pady=10) 

label = tk.Label(root, text="ยังไม่ได้คลิก", font=("Arial", 12))
label.pack(pady=10)


root.mainloop()
```
