```python
def flexi_calc(mode, *args):
    if mode == "add":  
       
        return sum(args)
    
    elif mode == "mult":  
        n = 1
        for b in args:
           
            n = n * b
        return n
    
    else:
        return 
print(f"บวก: {flexi_calc('add', 5, 2, 8)}")   
print(f"คูณ: {flexi_calc('mult', 20, 9, 5)}")
```

```python
def stat_calc(mode, *args):
   
    if not args:
        return 0 

    if mode == "avg":
        
        total_sum = sum(args)
        count = len(args)
        if count == 0:
            return 0
        return total_sum / count
        
    elif mode == "max":
       
        return max(args)
        
    else:
    
        print(f"Error: Unknown mode '{mode}'. Use 'avg' or 'max'.")
        return None
print(f'Average: {stat_calc("avg", 10, 20, 30)}') 
print(f'Maximum: {stat_calc("max", 5, 2, 8, 1)}')
print(f'Empty Maximum: {stat_calc("max")}') 
```
