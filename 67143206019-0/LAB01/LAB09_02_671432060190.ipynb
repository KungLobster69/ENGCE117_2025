```python
class Playlist:
    def __init__(self, songs):
       
        self.songs = songs
        
    def __len__(self):
        
        return len(self.songs)
    
    def __getitem__(self, index):
      
        if isinstance(index, slice):
            
            return Playlist(self.songs[index])
        
        
        return self.songs[index]

    def __setitem__(self, index, value):
        
        self.songs[index] = value

    def __contains__(self, item):
       
        return item in self.songs

    def __str__(self):
        
        return f"Playlist: {self.songs}"


my_Playlist = Playlist(["Song A", "Song B", "Song C", "Song D"])


new_p = my_Playlist[0:2]

print(f"ผลลัพธ์จากการ Slice: {new_p}") 
print(f"ชนิดของ Object ใหม่: {type(new_p)}") 

print(f"เพลงแรกใน Playlist ใหม่: {new_p[0]}")
```
