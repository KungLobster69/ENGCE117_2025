```python
def make_uppercase(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper


def make_bold(func):
    def wrapper():
        return f"<b>{func()}</b>"
    return wrapper


def make_italic(func):
    def wrapper():
        return f"<i>{func()}</i>"
    return wrapper



@make_bold
@make_italic
@make_uppercase
def get_hello():
    return "hello world"

print(get_hello())
```

```python
def to_upper(func):
    """ทำหน้าที่รับ string จากฟังก์ชันแล้วแปลงเป็นตัวพิมพ์ใหญ่ทั้งหมด"""
    def wrapper():
        result = func()
        return result.upper()
    return wrapper


def add_brackets(func):
    """ทำหน้าที่รับ string จากฟังก์ชันแล้วครอบด้วยวงเล็บ [...]"""
    def wrapper():
        result = func()
        return f"[{result}]"
    return wrapper



@add_brackets
@to_upper
def get_message():
    return "hello world"


print(get_message())
```
