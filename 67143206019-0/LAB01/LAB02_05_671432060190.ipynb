```python
names = ["alice", "bob", "charlie"]
upper_names = [name.upper() for name in names]
print(f"Upper: {upper_names}")
```

```python
prices = [100, 250, 80, 500, 120]
vat_prices = [p * 1.07 for p in prices if p > 200]
print(f"VAT Prices: {vat_prices}")
```

```python
currencies = {'USD': 35, 'EUR': 38, 'JPY': 0.25}
inverted_currencies = {str(value): key for key, value in currencies.items() if value > 1}
print(f"Inverted: {inverted_currencies}")
```
