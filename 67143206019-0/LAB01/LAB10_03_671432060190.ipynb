```python
import tkinter as tk
from tkinter import messagebox 

def submit():
    if accept_var.get(): 
        
        messagebox.showinfo("Success", f"ลงทะเบียนชื่อ: {name_var.get()} เรียบร้อย!")
    else:
        
        messagebox.showwarning("Warning", "กรุณายอมรับเงื่อนไขก่อนกด Submit!")

root = tk.Tk()
root.title("Registration Form")
root.geometry("300x250") 

name_var = tk.StringVar() 
accept_var = tk.BooleanVar()

tk.Label(root, text="Username:").pack(pady=(15, 0))
tk.Entry(root, textvariable=name_var).pack(pady=5)
tk.Checkbutton(root, text="I accept the terms", variable=accept_var).pack(pady=10)

tk.Button(root, text="Submit", command=submit).pack(pady=10)

root.mainloop()
```
