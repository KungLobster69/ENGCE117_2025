```python
def sum_rec(n):
    if n == 0:
        return 0
    else:
        return n + sum_rec(n - 1)
print(sum_rec(5))
 
```

```python
def reverse_str(s):
    if s == "" :
        return ""
    else:
        return s[-1] + reverse_str(s[:-1])
print(reverse_str("123"))
```

```python
def power(base,exp):
    if exp == 0:
        return 1
    else:
        return base * power(base, exp - 1)
print(power(2, 3))    
```
