```python
try:
    with open("note.txt", "w", encoding="utf-8") as file:

        nickname = "Kittikorn"
        number = "67143206019-0"
        
       
        content_line1 = f"Nickname: {nickname},{number}"
        
      
        content_line2 = f"\nNumber: {number}"
        
 
        file.write(content_line1)
        
       
        file.write(content_line2)

    content_display = content_line1 + content_line2
    
    print("'note.txt' ")
    print(f"'{content_display.strip()}'")
    
except IOError as e:
    print(f" {e}")
        
```

```python
favorite_food = input("ไก่ทอด ")
content_to_append = f"\nfavorite_food: {favorite_food}"

try:
    
    with open("note.txt", "a", encoding="utf-8") as file:
        
       
        file.write(content_to_append) 
    
    
    print(f" 'note.txt'")
    print(f"  {content_to_append.strip()}")
    
except IOError as e:
    print(f" {e}")
```

```python

try:
   
    with open("note.txt", "r", encoding="utf-8") as file:
        
      
        file_content = file.read() 
        
   

    print(" 'note.txt':")
    print(file_content)
    
    
except FileNotFoundError:
    print(" 'note.txt' ")
    
except IOError as e:
    print(f" {e}")
```
