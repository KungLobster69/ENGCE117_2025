```python
import tkinter as tk

root = tk.Tk()
root.title("Numpad")

display = tk.Entry(root, font=("Arial", 18), justify='right')
display.grid(row=0, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")


buttons = [
    '1', '2', '3',
    '4', '5', '6',
    '7', '8', '9'
]

row_val = 1 
col_val = 0 

for btn_text in buttons:
    tk.Button(root, text=btn_text, width=10, height=3).grid(row=row_val, column=col_val, padx=2, pady=2)
    
    col_val += 1 
    
    if col_val > 2:
        col_val = 0
        row_val += 1

root.mainloop()
```

```python
import tkinter as tk

root = tk.Tk()
root.title("Numpad")

display = tk.Entry(root, font=("Arial", 20), justify='right')
display.grid(row=0, column=0, columnspan=3, padx=10, pady=10, sticky="nsew")

numbers = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2)
]

for (text, r, c) in numbers:
    tk.Button(root, text=text, width=10, height=3).grid(row=r, column=c, padx=2, pady=2)

button_0 = tk.Button(root, text="0", height=3)
button_0.grid(row=4, column=0, columnspan=2, padx=2, pady=2, sticky="nsew")

button_dot = tk.Button(root, text=".", width=10, height=3)
button_dot.grid(row=4, column=2, padx=2, pady=2)

root.mainloop()
```
