```python
text = "apple banana apple orange banana apple"
word_list = text.split()
word_counts = Counter(word_list)
most_frequent_word = word_counts.most_common(1)
print(f"Output: {most_frequent_word}")
```

```python
from collections import defaultdict
import pprint
names = ["Alice", "Bob", "Charlie", "David", "Eve"]
grouped_names = defaultdict(list)
for name in names:
    length = len(name)
    grouped_names[length].append(name)
final_output = dict(grouped_names)
print(f"รายการชื่อ: {names}")
pprint.pprint(final_output)
```

```python
from collections import deque 
logs_input = ["Log1", "Log2", "Log3", "Log4", "Log5"]
log_window = deque(maxlen=3) 
for log in logs_input:
    log_window.append(log)
final_logs = list(log_window)
print(f"# Output: {final_logs}")
```
