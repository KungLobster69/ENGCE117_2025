```python
#Level 1
def to_upper(func):
    def wrapper():
        text = func()
        return text.upper()
    return wrapper
```

```python
#Level 2
def add_brackets(func):
    def wrapper():
        text = func()
        return f"[{text}]"
    return wrapper
```

```python
#Level 3
@add_brackets
@to_upper
def say():
    return "hello python"
print(say())
```
