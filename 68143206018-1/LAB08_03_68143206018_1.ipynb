```python
#Level 1
from abc import ABC, abstractmethod
class IoTDevice(ABC):
    @abstractmethod
    def turn_on(self):
        pass
    @abstractmethod
    def turn_off(self):
        pass
class SmartLight(IoTDevice):
    def turn_on(self):
        print("ไฟเปิด")
    def turn_off(self):
        print("ไฟปิด")
light = SmartLight()
light.turn_on()
light.turn_off()
```

```python
#Level 2
class SmartSpeaker(IoTDevice):
    def turn_on(self):
        print("ลำโพงเปิด")
try:
    speaker = SmartSpeaker()
except TypeError as e:
    print("Error:", e)
```

```python
#Level 3
def test_device(device):
    if isinstance(device, IoTDevice):
        print("Testing device...")
        device.turn_on()
        device.turn_off()
    else:
        print("ไม่ใช่อุปกรณ์ IoT ที่รองรับ")
light = SmartLight()
test_device(light)
```
