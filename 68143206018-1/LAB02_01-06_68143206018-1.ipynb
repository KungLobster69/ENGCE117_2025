```python
# Workshop 3. 1
inventory = {'apple': 10, 'banana': 20}
def check_quantity(item_name):
    return inventory.get(item_name, 0)
print(f"Check Durian: {check_quantity('durian')}")
```

```python
# Workshop 3. 2
inventory = {'apple': 10, 'banana': 20}
def add_stock(item, quantity):
    current_quantity = inventory.get(item, 0)
    new_quantity = current_quantity + quantity
    inventory[item] = new_quantity
add_stock('apple', 5)
add_stock('orange', 10)
print(inventory)
```

```python
# Workshop 3. 3
inventory = {'apple': 10, 'banana': 20}
branch2 = {'banana': 5, 'orange': 10}
for item, quantity in branch2.items():
    current_stock = inventory.get(item, 0)
    inventory[item] = current_stock + quantity
print(inventory)
```

```python
# Workshop 6. 1
from collections import Counter
text = "apple banana apple orange banana apple"
words = text.split()
word_counts = Counter(words)
top_word = word_counts.most_common(1)
print(top_word)
```

```python
# Workshop 6. 2
from collections import defaultdict
names = ["Alice", "Bob", "Charlie", "David", "Eve"]
grouped_by_length = defaultdict(list)
for name in names:
    length = len(name)
    grouped_by_length[length].append(name)
print(dict(grouped_by_length))
```

```python
# Workshop 6. 3
from collections import deque
recent_logs = deque(maxlen=3)
log_messages = ['Log1', 'Log2', 'Log3', 'Log4', 'Log5']
for log in log_messages:
    recent_logs.append(log)
print(list(recent_logs))
```
