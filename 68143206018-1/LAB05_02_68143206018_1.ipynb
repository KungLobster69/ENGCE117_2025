```python
#Level 1
class Robot:
    def __init__(self, name, model):
        self.name = name
        self.model = model
r1 = Robot("R-01", "Explorer")
print(f"Robot: {r1.name} ({r1.model})")
```

```python
#Level 2
class Robot:
    def __init__(self, name, model, battery=100):
        self.name = name
        self.model = model
        self.battery = battery
r1 = Robot("R-02", "Scout")
print(f"{r1.name} Battery: {r1.battery}%")
```

```python
#Level 3
class Robot:
    def __init__(self, name, model, battery=100):
        self.name = name
        self.model = model
        self.battery = battery
        if self.battery < 20:
            print(f"Warning: {self.name} has Low Battery!")
r1 = Robot("R-03", "Explorer", 10)
```
