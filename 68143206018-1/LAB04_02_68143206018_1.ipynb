```python
#Level 1
def even_gen(limit):
    for i in range(limit+1):
        if i % 2 == 0:
            yield i
for n in even_gen(10):
    print(n)
```

```python
#Level 2
def frange(start, stop, step):
    while start < stop:
        yield start
        start += step
for x in frange(0, 1, 0.5):
    print(x)
```

```python
#Level 3
#Gen 1
def gen_numbers():
    for i in range(1,6):
        yield i
```

```python
#Gen 2
def square_gen(numbers):
    for n in numbers:
        yield n*n
```

```python
nums = gen_numbers()
squares = square_gen(nums)

for s in squares:
    print(s)
```
