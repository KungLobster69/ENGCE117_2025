```python
#Level 1
class WiFi:
    def connect(self):
        print("Connected to WiFi network")
class Bluetooth:
    def connect(self):
        print("Paired with Bluetooth device")
wifi = WiFi()
bt = Bluetooth()
wifi.connect()
bt.connect()
```

```python
#Level 2
def initialize_system(device):
    print("System initializing...")
    device.connect()
wifi = WiFi()
bt = Bluetooth()
initialize_system(wifi)
initialize_system(bt)
```

```python
#Level 3
def initialize_safe(device):
    try:
        device.connect()
    except AttributeError:
        print("Error: เชื่อมต่อไม่ได้")
wifi = WiFi()
initialize_safe(wifi)
initialize_safe("สายแลนปลอม")
initialize_safe(123)
```
