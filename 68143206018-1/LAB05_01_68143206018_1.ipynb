```python
#Level 3
class Robot:
    pass

r1 = Robot()
r1.name = "Commander"
r1.battery = 100

r2 = Robot()
r2.name = "Commander"
r2.battery = 100

print("Is r1 the same as r2?", r1 is r2)
print("Address r1:", id(r1))
print("Address r2:", id(r2))
```
