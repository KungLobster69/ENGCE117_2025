```python
#Level 1
class Circle:
    def draw(self):
        print("วาดวงกลม")
class Square:
    def draw(self):
        print("วาดสี่เหลี่ยม")
c = Circle()
s = Square()
c.draw()
s.draw()
```

```python
#Level 2
class Circle:
    def draw(self):
        print("วาดวงกลม")
class Square:
    def draw(self):
        print("วาดสี่เหลี่ยม")
shapes = [Circle(), Square(), Circle(), Square()]
print("--- เริ่มการวาดแบบ Loop ---")
for item in shapes:
    item.draw()
```

```python
#Level 3
class Circle:
    def draw(self):
        print("วาดวงกลม")
class Square:
    def draw(self):
        print("วาดสี่เหลี่ยม")
class Triangle:
    def draw(self):
        print("วาดสามเหลี่ยม")
def render_screen(shape_list):
    print("--- Rendering shapes ---")
    for shape in shape_list:
        shape.draw()
shapes = [Circle(), Triangle(), Square()]
render_screen(shapes)
```
