```python
#Level 1
class Animal:
    def make_sound(self):
        print("... (Silence)")
```

```python
#Level 2
class Dog(Animal):
    def make_sound(self):
        print("Woof! Woof!")
class Cat(Animal):
    def make_sound(self):
        print("Meow~ Meow~")
```

```python
#Level 3
band = [Dog(), Cat(), Dog(), Animal()]
for animal in band:
    animal.make_sound()
```
