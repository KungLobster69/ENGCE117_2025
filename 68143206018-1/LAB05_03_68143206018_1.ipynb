```python
#Level 1
class RobotPart:
    category = "Mechanical"
    def __init__(self, name):
        self.name = name
p1 = RobotPart("Motor")
print(f"Part: {p1.name}, Type: {RobotPart.category}")
```

```python
#Level 2
class RobotPart:
    category = "Mechanical"
    total_count = 0
    def __init__(self, name):
        self.name = name
        RobotPart.total_count += 1
p1 = RobotPart("Motor")
p2 = RobotPart("Sensor")
print("Total parts in system:", RobotPart.total_count)
```

```python
#Level 3
class RobotPart:
    category = "Mechanical"
    total_count = 0
    def __init__(self, name):
        self.name = name
        RobotPart.total_count += 1
p1 = RobotPart("Motor")
p2 = RobotPart("Sensor")
p3 = RobotPart("Camera")
print("Is it shared?", p1.total_count == RobotPart.total_count)
```
