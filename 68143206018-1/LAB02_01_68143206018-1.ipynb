```python
def super_sum(*nums):
    total = 0
    for num in nums:
        total += num
    return total
numbers = input("กรอกตัวเลขหลายๆ ตัว (คั่นด้วยช่องว่าง): ")
nums = [int(x) for x in numbers.split()]
print("ผลรวม =", super_sum(*nums))
```

```python
def flexi_calc(mode, *args):
    if mode == "add":
        return sum(args)
    elif mode == "mult":
        result = 1
        for num in args:
            result *= num
        return result
    else:
        return "Invalid mode"
mode = input("เลือกโหมด (add/mult): ").strip()
numbers = input("กรอกตัวเลขหลายๆ ตัว (คั่นด้วยช่องว่าง): ")
nums = [int(x) for x in numbers.split()]
result = flexi_calc(mode, *nums)
print("ผลลัพธ์ =", result)
```

```python
def stat_calc(*args):
    if not args:
        return 0, 0
    avg_value = sum(args) / len(args)
    max_value = max(args)
    return avg_value, max_value
numbers = input("กรอกตัวเลขหลายๆ ตัว (คั่นด้วยช่องว่าง): ").strip()
if numbers == "":
    avg, max_val = stat_calc()
else:
    nums = [int(x) for x in numbers.split()]
    avg, max_val = stat_calc(*nums)
print("ค่าเฉลี่ย (avg) =", avg)
print("ค่ามากสุด (max) =", max_val)
```
