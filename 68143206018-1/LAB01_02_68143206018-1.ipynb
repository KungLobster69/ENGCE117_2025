```python
# Workshop  2. 2
def get_stats(numbers):
    return min(numbers), max(numbers)
numbers = [3, 1, 5, 4, 2]
min_val, max_val = get_stats(numbers)
print(f"Min: {min_val}, Max: {max_val}")
```
