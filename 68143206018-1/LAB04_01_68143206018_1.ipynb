```python
def sum_rec(n):
  if n == 1:
    return 1
  return n + sum_rec(n - 1)
print(f"Sumu (5): {sum_rec(5)}")
```

```python
def reverse_str(s):
    if len(s) <= 1:
        return s
    return s[-1] + reverse_str(s[:-1])
print(f"Reverse ('SRD'): {reverse_str('SRD')}")
```

```python
def power(base, exp):
    if exp == 0:
        return 1
    return base * power(base, exp - 1)
print(f"Power (2, 3): {power(2, 4)}")
```
