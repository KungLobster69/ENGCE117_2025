```python
class robot:
    def __init__(self, name, model, battery = 100):
        self.name = name
        self.model = model
        if battery < 20:
            print(f"Warning: {self.name} has Low Battery!")
        self.battery = battery
        

r1 = robot("R-01", "Explorer")
print(f"Robot: {r1.name} ({r1.model})")
```

```python
r2 = robot("R-02", "Explorer")
print(f"Robot: {r2.name} Battery: {r2.battery}%")
```

```python
r3 = robot("R-03", "Explorer", 15)
```
