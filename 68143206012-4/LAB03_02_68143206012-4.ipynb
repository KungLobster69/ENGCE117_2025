```python
raw_names = [" somchai ", "alice", " BOB", "c"]
map_names = list(map(lambda name: name.strip(' ').lower(), raw_names))
print(map_names)
```

```python
filter_names = list(filter(lambda name: len(name) >= 2, map_names))
print(filter_names)
```

```python
com_names = [name.strip(' ').lower() for name in raw_names if len(name) >= 2]
print(com_names)
```
