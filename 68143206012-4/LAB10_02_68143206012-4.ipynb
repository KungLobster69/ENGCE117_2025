```python
import tkinter as tk

def change_text(label):
    label.config(text="คลิกแล้วจ้า!", fg="red")

root = tk.Tk()
tk.Button(root, text="คลิกฉันสิ!", command=lambda: change_text(label)).pack()

label = tk.Label(root, text="ยังไม่ได้คลิก", font=(18))
label.pack()

root.title("My First App")
root.geometry("200x150")

root.mainloop()
```
