```python
def even_gen(limit):
    for i in range(limit + 1):
        if i % 2 == 0:
            yield i

for num in even_gen(10):
    print(num)
```

```python
def float_range(start, stop, step):
    current = float(start)
    while current < stop:
        yield current
        current += step

for val in float_range(0, 1, 0.5):
    print(val)
```

```python
def factory():
    for i in range(1, 6):
        yield i

def square_factory(num):
    for n in num:
        yield n ** 2

pipeline = square_factory(factory())

for r in pipeline:
    print(r)
```
