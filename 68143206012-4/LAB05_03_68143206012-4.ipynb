```python
class RobotPart:
    category = "Mechanical"
    total_count = 0
    def __init__(self, name):
        self.name = name
        RobotPart.total_count += 1

part1 = RobotPart("Motor")

print(f"Part: {part1.name} Type: {part1.category}")
```

```python
print(f"Total parts in system: {RobotPart.total_count}")
```

```python
part2 = RobotPart("servo")

print(f"Is it shared? {part1.total_count is RobotPart.total_count}")
```
