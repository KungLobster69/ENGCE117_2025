```python
inventory = {'apple': 10, 'banana': 20}
def check_price(item_name):
    item_num = inventory.get(item_name, 0)
    return item_num
item_check=check_price('durain')
print(f"Check : {item_check}")
```

```python
def add_stock(item, quantity):
    newdata = {item : inventory.get(item, 0) + quantity}
    return newdata
inventory.update(add_stock('apple', 5))
print(inventory)
```

```python
branch2 = {'banana': 5, 'orange': 10}
for item, qty in branch2.items():
    inventory[item] = inventory.get(item, 0) +qty
print(inventory)
```
