```python
#Level 1
class Playlist:
    def __init__(self, songs):
        self.songs = list(songs)  

    def __len__(self):
        return len(self.songs)    

    def __getitem__(self, index):
        #Level 3
        if isinstance(index, slice):
           
            return Playlist(self.songs[index])
        
       
        return self.songs[index]  

    #Level 2
    def __setitem__(self, index, value):
        self.songs[index] = value 

    def __contains__(self, song):
        return song in self.songs
        
    def __str__(self):
        return str(self.songs)

my_playlist = Playlist(["Song 1", "Song 2", "Song 3"])

print(f"Playlist: {my_playlist}")
print(f"จำนวนเพลง: {len(my_playlist)}") 
print(f"เพลงแรก: {my_playlist[0]}")      
print(f"เพลงสุดท้าย: {my_playlist[2]}")  


my_playlist[1] = "New Song 2" 
print(f"มี 'New Song 2' ไหม?: {'New Song 2' in my_playlist}")

sub_playlist = my_playlist[1:3]
print(f"Sub Playlist: {sub_playlist}")
print(f"Type: {type(sub_playlist)}")
```
