```python
class Square:
    def __init__(self, side):
        self.__side = side
    
    @property
    def side(self):
        return self.__side
    
    @side.setter
    def side(self, value):
        if value > 0:
            self.__side = value
        else:
            raise ValueError("ความยาวด้านต้องมากกว่า 0")
            
    @property
    def area(self):
        return self.__side * self.__side
    
s = Square(4)
print(f"Side: {s.side}, Area: {s.area}")

s.side = 5
print(f"New Area: {s.area}")

s.side = -10
#s.area = 100
```
