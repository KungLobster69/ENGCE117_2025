```python
def super_sum(*nums):
    print(sum(nums))

super_sum(1, 2, 3, 5, 6, 7, 8, 9)
```

```python
def flexi_calc(mode, *args):
    if mode == 'add':
        print(sum(args))
    if mode == 'mult':
        total = 1
        for num in args:
            total *= num
        print(total)

flexi_calc('add', 2, 3)
flexi_calc('mult', 4, 7)
```

```python
def stat_calc(mode, *args):
    if args != None :
        if mode == 'avg':
            print(sum(args)/len(args))
        if mode == 'max':
            print(max(args))
    else: return 0 

stat_calc("avg", 10, 20, 30)
stat_calc("avg", 0)
stat_calc("max", 1, 4, 8)
```
