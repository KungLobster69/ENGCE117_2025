```python
class Character:
    def __init__(self, name, hp, power):
        self.name = name
        self.hp = hp
        self.power = power
        
    def take_damage(self, amount):
        self.hp -= amount
        if self.hp < 0:
            self.hp = 0
        print(f"{self.name} took damage. (â¤ï¸HP left: {self.hp})")

    def attack(self, target):
        self.atk = self.power
        print(f"{self.name} attacks âš”ï¸ {target.name}!")
        target.take_damage(self.atk)

    def is_alive(self):
        return (self.hp > 0)
```

```python
class Warrior(Character):
    def __init__(self, name, hp, power, defense):
        super().__init__(name, hp, power)
        self.defense = defense
        
    def take_damage(self, amount):
        damage = amount - self. defense
        if damage < 1:
            damage = 1
        super().take_damage(damage)

class Mage(Character):
    def __init__(self, name, hp, power, mana):
        super().__init__(name, hp, power)
        self.mana = mana
        
    def attack(self, target):
        self.atk = self.power + self.mana
        print(f"{self.name} cast Fireball ðŸ”¥ at {target.name}!")
        target.take_damage(self.atk)
```

```python
arthur = Warrior("Arthur", 100, 20, 40)
loki = Mage("Loki", 100, 20, 40)

def battle(p1, p2):
    print("--Battle Start--")
    while p1.is_alive() and p2.is_alive():
        p1.attack(p2)
        if not p2.is_alive():
            break
        p2.attack(p1)

    winner = p1.name if p1.is_alive() else p2.name
    print(f"ðŸ† Winner: {winner}")

battle(arthur, loki)
```
