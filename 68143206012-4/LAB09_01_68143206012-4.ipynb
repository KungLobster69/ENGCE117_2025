```python
#Level 1
class RPGCharacter :
    def __init__(self, name, hp, mp):
        self.name = name
        self.hp = hp
        self.mp = mp
        
    def __str__(self):
        return f"{self.name} (HP: {self.hp}, MP: {self.mp})"
    
    #Level 2    
    def __add__(self, other):
        if isinstance(other, RPGCharacter):
            return RPGCharacter(self.name + other.name, self.hp + other.hp, self.mp + other.mp)
        elif isinstance(other, int):
            #Level 3
            return RPGCharacter(self.name, self.hp + other, self.mp)
        else:
            return NotImplementedError
 
#Answer 1 
char1 = RPGCharacter("Hero", 100, 50)
print(char1)

char2 = RPGCharacter("Mage", 50, 100)
print(char2)

#Answer 2
print(char1 + char2)

#Answer 3
print(char1 + 50)

print(char1 + "50")
```
