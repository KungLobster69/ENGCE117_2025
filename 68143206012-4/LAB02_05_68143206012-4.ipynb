```python
names = ["alice", "bob", "charlie"]
up_name = [a.upper() for a in names]
print("Upper: ",up_name)
```

```python
prices = [100, 250, 80, 500, 120]
vat_prices = [x + (x*0.07) for x in prices if x > 200]
print("VAT Prices: ",vat_prices)
```

```python
currencies = {'USD': 35, 'EUR': 38, 'JPY': 0.25}
inverted = {c : v for v, c in currencies.items() if c > 1}
print("Inverted: ",inverted)
```
