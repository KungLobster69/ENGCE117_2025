```python
class Square:
    def __init__(self,s):
        self.__side = s

    @property
    def side(self):
        return self.__side
    @side.setter
    def side(self,val): 
        if val >= 0:
            self.__side = val
        else:
            raise ValueError("Val =< 0")

    @property
    def area(self):
        return self.__side * self.__side
```

```python
s = Square(4)
print(f"Side:{s.side},Area:{s.area}")
```

```python
s.side = 5
print(f"New Area: {s.area}")
```

```python
s.side = -10
```

```python
s.area = 100
```
