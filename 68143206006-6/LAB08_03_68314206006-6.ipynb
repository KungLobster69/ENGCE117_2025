```python
from abc import ABC, abstractmethod
```

```python
class IoTDevice(ABC) :
    @abstractmethod
    def turn_on(self):
        pass
    @abstractmethod
    def turn_off(self):
        pass
```

```python
def restart(obj):
        obj.turn_on()
        obj.turn_off()
```

```python
class SmartLight(IoTDevice):
    def turn_on(self):
        print("üí° ‡πÑ‡∏ü‡πÄ‡∏õ‡∏¥‡∏î")
    def turn_off(self):
        print("üåë ‡πÑ‡∏ü‡∏õ‡∏¥‡∏î")
```

```python
lightbulb = SmartLight()

restart(lightbulb)
```

```python
class SmartSpeaker(IoTDevice) :
        def turn_on(self):
            print("üîä ‡∏•‡∏≥‡πÇ‡∏û‡∏á‡πÄ‡∏õ‡∏¥‡∏î")
```

```python
Speaker = SmartSpeaker()

try :
    restart(Speaker)
except TypeError : "‚ùå Error: Can't instantiate abstract class SmartSpeaker with abstract method turn_off"
```

```python
def test_device(device):
    print(f"Testing {device.__class__.__name__}...")
    if isinstance(device, IoTDevice):
        restart(device)
    else:
        print("‚ö†Ô∏è ‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡∏≠‡∏∏‡∏õ‡∏Å‡∏£‡∏ì‡πå IoT ‡∏ó‡∏µ‡πà‡∏£‡∏≠‡∏á‡∏£‡∏±‡∏ö")
```
