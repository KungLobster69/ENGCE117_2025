```python
import tkinter as tk
```

```python
def subm():
       usern = use_var.get()
       if var_acc.get() and usern: 
              label.config(text=f"Registration Success! User: {usern}", fg="black")
       elif not var_acc.get():
              label.config(text="You must accept ToS", fg="red")
       else:
              label.config(text="Please Enter Your User Name", fg="red")
```

```python
root = tk.Tk()

root.title("Registration Form")
root.geometry("300x200")

label = tk.Label(root, text="Type Your Username")
label.pack(pady=10)

use_var = tk.StringVar()
use_var = tk.Entry(root, width=40)
use_var.pack(padx=10, pady=10)


var_acc = tk.BooleanVar()
chk_b = tk.Checkbutton(root, text="I accept the terms of use", variable=var_acc)
chk_b.pack()

btn = tk.Button(root, text="Submit", command=subm)
btn.pack(pady = 5)



root.mainloop()
```
