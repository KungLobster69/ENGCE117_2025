```python
def to_upper(func):
    def wrapper():
        original = func()
        return original.upper()
    return wrapper
```

```python
@to_upper
def upper_txt():
    return "hello python"
```

```python
upper_txt()
```

```python
def add_bracket(func):
    def wrapper():
        result = func()
        return f"[{result}]"
    return wrapper
```

```python
    @add_bracket
    def bracket_txt():
        return "hello python"
```

```python
bracket_txt()
```

```python
@add_bracket
@to_upper
def stack_txt():
    return "hello python"

stack_txt()
```
