```python
class RpgCharac():
    def __init__(self, name, hp, mp):
        self.name = name
        self.hp = hp
        self.mp = mp

    def __str__(self):
        return f"{self.name}(HP:{self.hp} MP:{self.mp})"
    
    def __add__(self, other):
        if isinstance(other, int):
            self.hp += other
            return self
        elif isinstance(other, RpgCharac):
            return RpgCharac(self.name + " " + other.name, 
                             self.hp + other.hp, 
                             self.mp + other.mp
                             )
        return NotImplemented
```

```python
char1 = RpgCharac("Novice1", 100, 50)
print(f"{char1}")
```

```python
char2 = RpgCharac("Novice2", 100, 50)
char3 = char1 + char2
print(f"{char3}")
```
