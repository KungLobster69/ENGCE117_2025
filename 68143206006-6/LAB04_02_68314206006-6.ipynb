```python
def intermediate(start,step,stop):
    while start < stop+step:
        yield start
        start += step
```

```python
for i in intermediate(0,0.25,5):
    print(i)
```

```python
def gen1(limit):
    for i in range(1,limit+1):
        yield i      
```

```python
def gen2(val):
    for i in gen1(val):
        print(f"{i} PowerBy 2  = {i**2}")
```

```python
gen2(5)
```
