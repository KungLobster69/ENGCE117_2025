```python
class PlayList():
    def __init__(self, song_list, name = None):
        if name is None:
            self.name = self.name = f"Playlist_{id(self)}"
            self._item = song_list
      
    def __str__(self):
        return f"{self.name}:{self._item}"
    
    def __len__(self):
        return len(self._item)
    
    
    def __getitem__(self, order):
        if isinstance(order,slice):
            return PlayList(self.name,self._item[order])
        else :
            return self._item[order]
    
    def __setitem__(self, order, song_list):
        self._item[order] = song_list
    
    def __contains__(self, target):
        for item in self._item:
            if item == target:
                print(f"{target} is in the {self.name}")
                return None
        print(f"{target} is not in the {self.name}")
        return None   
    
    
```

```python
pl1 = PlayList(["song A", "song B", "song C", "song D"])
print(f"{pl1}")
```
