```python
class RobotPart:
    total_count = 0
    def __init__(self,Part,Category = "Mechanical"):
        RobotPart.total_count += 1
        self.Part = Part
        self.type = Category
        
```

```python
p1 = RobotPart("Motor")
print(f"Part:{p1.Part},Type:{p1.type}")
```

```python
p2 = RobotPart("Wings")
print(f"Totalpart in system:{RobotPart.total_count}")
```

```python
print(f"Check count by p1:{p1.total_count}")
print(f"Check count by MainClass:{RobotPart.total_count}")
#เพราะทั้งสองคำสั่งชี้ไปที่ตัวแปรแดียวกัน
```
