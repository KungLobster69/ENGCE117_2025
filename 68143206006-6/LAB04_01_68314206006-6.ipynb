```python
def sum_rec(n):
    if n <= 0:
        return n
    return n + sum_rec(n-1)
    print(n)
```

```python
sum_rec(3)
```

```python
def reverse_str(s: str):
    if len(s) <= 1:
        return s
    return s[-1] + reverse_str(s[:-1])
```

```python
reverse_str("Jelly")
```

```python
def power(base,exp):
    if exp == 0:return 1
    if base == 0:return 0
    
    return power(base,exp-1) * base
```

```python
power(-6,2)
```
