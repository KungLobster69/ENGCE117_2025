```python
class Robot:
    def __init__(self,name,model,battery=100):
        self.name = name
        self.model = model
        self.battery = battery
        self.battery_state()
    def battery_state(self):
        if self.battery <= 20:
            print(f"Warning:{self.name} has Low Battery!")
```

```python
r1 = Robot("R-01","Explorer")
print(f"Robot:{r1.name},({r1.model})")
```

```python
r2 = Robot("R-02","Explorer",95)
print(f"Robot:{r2.name},({r2.model}) Battery:{r2.battery}%")
```

```python
r3 = Robot("R-03", "Explorer", 20)

print(f"Robot: {r3.name}, ({r3.model}) Battery: {r3.battery}%")
```
