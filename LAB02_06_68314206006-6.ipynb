```python
from collections import Counter,defaultdict,deque
```

```python

text = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']
c = Counter(text)
mc = c.most_common(1)
print(mc)
```

```python
names = ["Alice", "Bob", "Charlie", "David", "Eve"]
group = defaultdict(list)
for name in names : group[len(name)].append(name)
print(dict(group))
```

```python
logs = deque(maxlen=3)
for i in range(1, 6):
    message = input(f"Enter Log {i}: ")
    logs.append(message)
print("Last 3 Logs:", list(logs))
```
