```python
# Workshop 1: The Universal Calculator
# Level 1: Beginner (Super Sum)

def super_sum(*nums):
    total = 0
    for num in nums:
        total += num
    return total
print(super_sum(1, 2, 3, 4, 5))
```

```python
# Level 2: Intermediate (Flexi Calc)

def flexi_calc(mode, *args):
    if mode == "add":
        result = 0
        for num in args:
            result += num
        return result
    elif mode == "mult":
        result = 1
        for num in args:
            result *= num
        return result
    else:
        return "Invalid mode"
print(flexi_calc("add", 1, 2, 3))      
print(flexi_calc("mult", 1, 2, 3, 4))  
```

```python
# Level 3: Challenge (Stats Master)

def stat_calc(mode, *args):
    if len(args) == 0:
        return 0
    if mode == "avg":
        return sum(args) / len(args)
    elif mode == "max":
        return max(args)
print(stat_calc("avg", 10, 20, 30))  
print(stat_calc("avg"))               
```
