```python
# Workshop 3 - Inventory Manager
#Level 1 (Easy): การเข้าถึงและตรวจสอบ (Safe Access)

inventory = {'apple': 10, 'banana': 20}
def check_quantity(item_name):
    return inventory.get(item_name, 0)
print(f"Check Durian: {check_quantity('durian')}")
```

```python
# Level 2 (Medium): การอัปเดตข้อมูล (Smart Update)

inventory = {'apple': 10, 'banana': 20}
def add_stock(item, quantity):
    current_quantity = inventory.get(item, 0)
    new_quantity = current_quantity + quantity
    inventory[item] = new_quantity
add_stock('apple', 5)
add_stock('orange', 10)
print(inventory)
```

```python
# Level 3 (Hard): การรวมสาขา (Merging with Logic)

inventory = {'apple': 10, 'banana': 20} 
branch2 = {'banana': 5, 'orange': 10}
for item, quantity in branch2.items():
    current_stock = inventory.get(item, 0)
    inventory[item] = current_stock + quantity
print(inventory)
```
