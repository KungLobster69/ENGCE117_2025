```python
# Workshop 3: My First File
# Level 1: Beginner (สร้างไฟล์ใหม่)

with open('note.txt', 'w') as file:
    file.write('Nickname: Jo\n')
    file.write('Nickname: CHIN\n')
```

```python
# Level 2: Intermediate (เขียนต่อท้าย)

favorite_food = input("Enter your favorite food: ")
with open('note.txt', 'a', encoding='utf-8') as f:
    f.write("\nFavorite Food: " + favorite_food)

print("Appendedto note.txt.")
```

```python
# Level 3: Challenge (อ่านข้อมูล)

with open('note.txt', 'r', encoding='utf-8') as f:
    content = f.read()
print("\n--- Content of note.txt ---")
print(content)
```
