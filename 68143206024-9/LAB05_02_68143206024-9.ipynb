```python
# Workshop 2: Robot Initialization
# Level 1 (Easy): จงแก้ไขคลาส Robot ให้มี__init__ ที่รับพารามิเตอร์ name และ model

class Robot:
    def __init__(self, name, model):
        self.name = name
        self.model = model
r1 = Robot("R-01", "Explorer")
print(f"Robot: {r1.name}, Model: {r1.model}")
```

```python
# Level 2 (Medium): เพิ่มพารามิเตอร์ battery โดยกำหนดค่า Default เป็น 100 และสร้าง Object โดยไม่ส่งค่าแบตเตอรี่

r2 = Robot("R-02", "Worker")
r2.battery = 100  
print(f"Robot: {r2.name} Battery: {r2.battery}%")
```

```python
# Level 3 (Hard): เพิ่มพิ่ Logic ใน __init__ เพื่อตรวจสอบว่าหาก batteryต่ำกว่า 20 ให้พิมพ์คำเตือน "Warning: Low Battery"

class Robot:
    def __init__(self, name, model, battery=100):
        self.name = name
        self.model = model
        self.battery = battery
        if self.battery < 20:
            print(f"Warning: {self.name} has Low Battery!")
r3 = Robot("R-03", "Scout", battery=15)
```
