```python
# Workshop 1: Coding The Robot Factory
# Level 1 (Easy) Defining Structure: จงเขียนโค้ดเพื่อสร้างคลาสว่างชื่อ Robot และสร้างวัตถุชื่อ r1

class Robot:
    pass
r1 = Robot()
print(r1)
```

```python
# Level 2 (Medium) Dynamic Assignment:

class Robot:
    pass
r1 = Robot()
r1.name = "Commander"
r1.battery = 100
print(f"Robot Name: {r1.name}, Battery: {r1.battery}%")
```

```python
# Level 3 (Hard) Refenrence Analysis: 

class Robot:
    pass
r2 = Robot()
r2.name = "Commander"
r2.battery = 100
print(f"Is r1 the same as r2? {r1 is r2}")
```
