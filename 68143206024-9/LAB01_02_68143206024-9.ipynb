```python
# Workshop 2 - The Tuple Master
# Level 1 (Easy): การกระจายค่าพื้นฐาน (Basic Unpacking)

student = ("Somchai","Engineering", 3.75)
name, debt, gpa = student
print("Name:", name)
print("Debt:", debt)
print("GPA:", gpa)
```

```python
# Level 2 (Medium): ฟังก์ชันคืนค่าหลายตัว (Multiple Return)

def get_stats(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    return (min_val, max_val)
numbers = [1, 5]
min_value, max_value = get_stats(numbers)
print("Min:", min_value)
print("Max:", max_value)
```

```python
# Level 3 (Hard): การจัดการข้อมูลส่วนเกิน (Extended Unpacking)

scores = [10, 20, 30,40, 50, 60]
low, *other, high = scores
print(f"Low: {low}, high: {high}, Other: {other}")
```
