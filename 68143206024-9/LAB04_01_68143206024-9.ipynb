```python
# # Workshop 1 - The Recursion Master
# Level 1: Beginner (ผลรวมเลข)

basecase = 1
def sum_rec(n):
    if n == basecase:
        return 1
    else:
        return n + sum_rec(n - 1) 

print(sum_rec(5))
```

```python
# level 2: intermediate (กลับหลังหัน)

def reverse_str(s):
    if len(s) == 0:
        return s
    else:
        return s[-1] + reverse_str(s[:-1])
print(reverse_str("ABC"))
```

```python
# level 3: Challenger (เลขยกกำลัง)

def power(base, exp):
    if exp == 0:
        return 1
    else:
        return base * power(base, exp - 1)
print(power(2, 3)) 
print(power(5, 0)) 
```
