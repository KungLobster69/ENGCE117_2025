```python
# workshop 1 - The Slicing Master
# Level 1 (Easy): การตัดคำพื้นฐาน

scores = [80, 55, 90, 75, 60]
first_two_scores = scores[:2]
last_score = scores[-1]
print(first_two_scores, last_score)

```

```python
# Level 2 (Medium): การจัดการลำดับ

ids = [101, 102, 103, 104, 105, 106]
first_three_ids = ids[::-1]
last_two_ids = ids[::-2]
print(last_two_ids)
```
