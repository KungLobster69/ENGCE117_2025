```python
# Workshop 4 - Social Circle

visitors = ['user1', 'user2', 'user1', 'user3', 'user2']
unique_visitors = set(visitors)
print("Unique Count:", len(unique_visitors))
```

```python
# Level 2 (Medium): การหาจุดร่วม (Common Interests)

group_A = {'Reading', 'Coding', 'Gaming'}
group_B = {'Coding', 'Traveling', 'Gaming'}
common_interests = group_A & group_B
print(f"Common: {common_interests}")
```

```python
# Level 3 (Hard): การหาจุดต่างเฉพาะกลุ่ม (Exclusive)

group_A = {'Reading', 'Coding', 'Gaming'}
group_B = {'Coding', 'Traveling', 'Gaming'}
only_A = group_A - group_B
diff = group_A ^ group_B
print(f"Only A: {only_A}")
print(f"Diff: {diff}")
```
