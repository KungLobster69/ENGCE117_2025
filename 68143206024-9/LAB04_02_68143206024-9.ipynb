```python
# Workshop 2 - The Generator Wizard
# Level 1: Beginner (เลขคู่)

def even_gen(limit):
    for num in range(limit):
        if num % 2 == 0:
            yield num
for even_number in even_gen(10):
    print(even_number)
```

```python
# Level 2: Intermediate (ช่วงทศนิยม)

def frange(start, end, step):
    current = start
    while current < end:
        yield current
        current += step
for num in frange(0, 1, 0.5):
    print(num)
```

```python
# Level 3: Challenge (ต่อท่อประมวลผล)

def gen1():
    for i in range(1, 6):
        yield i
def gen2(input_gen):
    for value in input_gen:
        yield value ** 2
pipeline = gen2(gen1())
for result in pipeline:
    print(result)
```
