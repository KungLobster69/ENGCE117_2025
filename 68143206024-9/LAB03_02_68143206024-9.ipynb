```python
# Workshop 2: The Data Cleaner
# Level 1: Beginner (Cleaner)

raw_names = [" SOMCHAI ","ALICE"," BOB",]
cleaned_names = list(map(lambda x: x.strip().lower(), raw_names))
print(cleaned_names)
```

```python
# Level 2: Intermediate (Filter)
# โจทย์:ต่อจาก Level 1 ให้ใช้ filter เพื่อตัดชื่อที่สั้นเกินไป (ความยาว <= 2)

raw_names = [" SOMCHAI ","ALICE"," BOB","C"]
filtered_names = list(filter(lambda x: len(x) > 2, cleaned_names))
print(filtered_names)
```

```python
# Level 3: Challenge (One-Liner)
# โจทย์:เขียนโค้ดทั้งหมด (Clean + Filter) ให้จบในบรรทัดเดียวโดยใช้List Comprehension

final_names = [name.strip().lower() for name in raw_names if len(name.strip()) > 2]
print(final_names)
```
