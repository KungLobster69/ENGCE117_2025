```python
# Workshop 3: Robot Inventory System
# Level 1 (Easy) - Shared Property: จงสร้างคลาส RobotPart ที่มี Class Attribute ชื่อ category = "Mechanical" และรับname ผ่าน __init__

class RobotPart:
    category = "Mechanical"
    def __init__(self, name):
        self.name = name
part1 = RobotPart("Motor")
print(f"Part: {part1.name}, Type: {RobotPart.category}")
```

```python
#Level 2 (Medium) - Centralized Counter: เพิ่มพิ่ Class Attribute total_count = 0 และเขียให้บวกค่าขึ้น 1 ทุกครั้งที่มีการสร้างชิ้นส่วนใหม่

class RobotPart:
    category = "Mechanical"
    total_count = 0
    def __init__(self, name):
        self.name = name
        RobotPart.total_count += 1
part1 = RobotPart("Motor")
part2 = RobotPart("Sensor")
print(f"total parts in system: {RobotPart.total_count}")
```

```python
# Level 3 (Hard) - Shared Context Verification

class RobotPart:
    pass
    category = "Mechanical"
    total_count = 0

p1 = RobotPart()
p2 = RobotPart()

print("Is it shared?", p1.total_count == RobotPart.total_count)
```
