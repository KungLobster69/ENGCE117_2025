```python
#Lavel 1
def sum_rec(n):
    if n == 1:       #ถ้าเหลือแค่ 1 ก็คืนค่า 1
        return 1
    elif n == 0:    #กันค่าเป็นศูนย์
        return 0
    else:
        return n + sum_rec(n - 1)  #n + ผลรวมของ n-1
sum_rec(20)
```

```python
#Lavel 2
def reverse_str(s):
    if len(s) <= 1:  #ถ้าความยาวของ s น้อยกว่ารึเท่ากับ คืนค่านั้นเลย
        return s
    else:
        return s[-1] + reverse_str(s[:-1])   #ตัวสุดท้ายบวกกับตัวที่เหลือก่อนหน้า
reverse_str('ABC')
```

```python
#Lavel 3
def power(base, exp):
    if exp == 0:  #ถ้าเลขยกกำลังเป็น 0 คืนค่า1 ตามหลักคณิตศาสตร์
        return 1
    else:
        return base * power(base, exp - 1)   #เอาฐานคูณกับเลขยกกำลังที่ลดลงทีละ 1
power(2,3)
```
