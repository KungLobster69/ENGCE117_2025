```python
#Lavel 1
raw_names = [' somchai','alice','BOB','c']

low_names = list(map(lambda x: x.lower().strip(), raw_names))
print(low_names)
```

```python
#Lavel 2
cut_names = list(filter(lambda n: len(n) <= 2, raw_names))

new = list(set(raw_names) - set(cut_names))

print(new)
```
