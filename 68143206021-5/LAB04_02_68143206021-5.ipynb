```python
#Lavel 1
def even_gen(limit):
    for i in range(limit + 1):
        if i % 2 == 0:  #ถ้า i เป็นเลขคู่ yield ส่งค่าออกไปทีละตัว
            yield i

for num in even_gen(10):
    print(num)
```

```python
#Lavel 2
def frange(start, stop, step):
    while start < stop:
        yield start, 10
        start += step

for x in frange(0, 1, 0.2):
    print(x)
```

```python
#Lavel 3
def gen1():
    for i in range(1, 6):
        yield i                #สร้างเลข1-5 แล้วส่งไปที่ gen2 ทีละตัว

def gen2(numbers):
    for num in numbers:   #รับเลขมาแล้วแปลงเป็นกำลังสอง
        yield num ** 2

for result in gen2(gen1()):  #วนลูปแสดงผล gen1 gen2
    print(result)
```
