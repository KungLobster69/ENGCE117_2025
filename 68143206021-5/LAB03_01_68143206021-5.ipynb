```python
#Lavel 1
def super_sum(*num):
    for n in num:
        return sum(num)
    
print(f'Output:',super_sum(1 ,2, 3))#ส่งตัวเลขเข้าไปใน Function
```

```python
#Lavel 2 
def flexi_cale(mode, *args):
    if mode == 'add':
        return sum(args)
    elif mode == 'mult':
        result = 1
        for n in args:
            result *= n
        return result
    else:
        raise ValueError("Unsuported mode. Use 'add' or 'mult'.")

print(flexi_cale('add', 1, 2, 3, 4, 5))
print(flexi_cale('mult', 2, 3))
```

```python
#Lavel 3
def stat_calc(mode, *args):
    
    #จัดการ args ว่าง (คืนค่า 0)
    if len(args) == 0:
        return 0

    if mode == "avg":
        return sum(args) / len(args) #คำนวณค่าเฉลี่ย (ผลรวม / จำนวน)
    
    elif mode == "max":
        return max(args)
    
    else:
        return None 
        
print('ค่าเฉลี่ย:',stat_calc("avg", 10, 20, 30))
print(stat_calc("avg"))
print('ค่ามากสุด:',stat_calc("max", 10, 20, 30, 40))
```
