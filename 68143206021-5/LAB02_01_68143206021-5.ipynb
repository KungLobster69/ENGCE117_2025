```python
#Lavel 1 การตัดคำพื้นฐาน
scores = [80, 55, 90, 75, 60]

data = scores.pop()

print(f'Output: {scores[:2]} {data}')
```

```python
#Lavel 2 การจัดการลำดับ
ids = [101, 102, 103, 104, 105, 106]
del ids[::2] #ลบ index ทั้งหมด โดยลบเ index เว้น index
print(f'Output: {ids[::-1]}')
```

```python
#Lavel 3 แก้ไขข้อมูลแบบกลุ่ม
data = [10, 20, 99, 99, 99, 60, 70]

data[2:5] = [0] #แทนที่ 0 ที่ index 2-4

print(f'Output: {data}')
```
