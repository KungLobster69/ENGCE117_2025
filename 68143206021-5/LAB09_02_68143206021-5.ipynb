```python
class Playlist:
    def __init__(self, songs):
        #songs เป็น list ของชื่อเพลง
        self.songs = list(songs)

    def __len__(self):
        #คืนค่าจำนวนเพลงทั้งหมด
        return len(self.songs)

           #เข้าถึงเพลงด้วย index
    def __getitem__(self, index):
        if isinstance(index, slice):
            return Playlist(self.songs[index])
        return self.songs[index]


        #แก้ไขชื่อเพลง
    def __setitem__(self, index, value):
        self.songs[index] = value

    def __contains__(self, item):
        #ตรวจสอบว่ามีเพลงนี้ไหม
        return item in self.songs

    def __str__(self):
        return "Playlist: " + ", ".join(self.songs)



    #สร้าง Playlist
    p = Playlist(["Song A", "Song B", "Song C", "Song D"])
    print("Initial:", p)

    #ตรวจสอบความยาวและเข้าถึงเพลง
    print("Length:", len(p))      
    print("First song:", p[0])  
    print("Last song:", p[3])      

    #ก้ไขเพลง
    p[1] = "New Song B"
    print("After edit:", p)       

    # ตรวจสอบเพลงด้วย index
    print("'Song C' in p?", "Song C" in p)
    print("'Song X' in p?", "Song X" in p)

    sub_p = p[0:2]
    print("Sub playlist:", sub_p)  
```
