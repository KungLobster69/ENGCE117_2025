```python
class Animal:
    def make_sound(self):
        print('...Silence')

class Dog(Animal):
    def make_sound(self):
        print('woof! Woof!')

class Cat(Animal):
    def make_sound(self):
        print('Meow~ Meow~')

band = [Dog(), Cat(), Dog(), Animal()]

for x in band:
    x.make_sound()
```
