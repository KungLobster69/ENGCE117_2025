```python
class square():
    def __init__(self, side):
        self.__side = side

    @property
    def side(self):
        return self.__side

    @side.setter
    def side(self, new):
        if new > 0:
            self.__side = new
        else:
            raise ValueError('ต้องมากกว่า 0')

    @property
    def area(self):
        return self.__side * self.__side

s = square(4)
print(f'Side: {s.side}, Area: {s.area}')
s.side = 5
print(f"New Area = {s.area}")
s.side = -10
print(f"New Area = {s.area}")

```
