```python
import tkinter as tk
def on_button_click():
    label.config(text="Button was clicked!!", fg="red")
root = tk.Tk()
root.title("My Second App")
label = tk.Label(root, text="The button wasn't click yet!?")
label.pack(pady=10)
button = tk.Button(root, text="Click Me!!", command=on_button_click)
button.pack(pady=10)
root.geometry("200x150")
root.mainloop()
```
