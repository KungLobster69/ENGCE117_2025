```python
#Lavel 1
class circle:
    def draw(self):
        print(" วาดวงกลม")

class square:
    def draw(self):
        print(" วาดสี่เหลี่ยม")

C = circle()
S = square()

C.draw()
S.draw()
```

```python
#Lavel 2
shapes = [circle(), square(), circle(), square()]
for i in shapes:
    i.draw()
```

```python
#Lavel 3
def reder_shape(shape_list):
    for i in shape_list:
        i.draw()

class tiangle:
    def draw(self):
        print(" วาดสามเหลี่ยม")

reder = [circle(), tiangle(), square()]

print('--- Rendering 3 shapes ---')
reder_shape(reder)
```
