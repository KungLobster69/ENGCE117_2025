```python
# Level 1 แปลงข้อความเป็นตัวพิมพ์ใหญ่
def to_upper(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper

# Level 2
def add_brackets(func):
    def wrapper():
        result = func()
        return f"[{result}]"
    return wrapper

@add_brackets
@to_upper
def get_text():
    return "hello python"


print("Result:", get_text())
```
