```python
n = int(input("กรุณาใส่ตัวเลข n: "))

def sum_rec(n):
    if n == 0:
        return 0
    else:
        return n + sum_rec(n - 1)

print(sum_rec(n))
```

```python
base = int(input("กรุณาใส่ตัวเลขฐาน: "))
exp = int(input("กรุณาใส่เลขชี้กำลัง: "))

def power_rec(a, n):
    if n == 0:
        return 1
    elif n < 0:
        return 1 / power_rec(a, -n)
    else:
        return a * power_rec(a, n - 1)
result = power_rec(base, exp)
print(f"{base} ยกกำลัง {exp} = {result}")
```
