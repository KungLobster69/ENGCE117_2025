```python
inventory = {'apple': 10,'banana': 20}
def check_price(item_name):
    return inventory.get(item_name, 0)
print(f"Chack: Durian {check_price('durian')}")
```

```python
inventory = {'apple': 10,'banana': 20}
def add_stock(item, quantity) :
    inventory[item] = inventory.get(item, 0 ) + quantity
add_stock('apple', 5)
print(inventory)
```

```python
inventory = {'apple': 10,'banana': 20}
branch2 = {'banana':5, 'orange':10}
def pules(inventory, branch2) :
    for item, qty in branch2.items():
        inventory[item] = inventory.get(item, 0) + qty
pules(inventory, branch2)
print(inventory)
```
