```python
text = "apple banana apple orange banana apple"
from collections import Counter
words = text.split()
counts = Counter(words)
common = counts.most_common(1)
print(f"Result:{common}")
```

```python
names = ["Alice", "Bob", "Charlie", "David", "Eve"]
from collections import defaultdict
names = defaultdict(list)
names['Alice','David'] = 5
names['Bob', 'Eve'] = 3
names['Charlie'] = 7
a = {num:name for name,num in names.items()}
print(f"Output:{a}")
```

```python
from collections import deque
log_system = deque(maxlen=3)
inputs = ['Log1', 'Log2', 'Log3', 'Log4', 'Log5']
for item in inputs:
    log_system.append(item)
    print(f"add {item}: {list(log_system)}")
print(f"(Output){list(log_system)}")
```
