```python
class Square:
    def __init__(self, side):
        if side <= 0:
            raise ValueError("ความยาวด้านต้องมากกว่า 0")
        self.__side = side  # Private

    # Property side (Getter)
    @property
    def side(self):
        return self.__side

    # Property side (Setter)
    @side.setter
    def side(self, value):
        if value <= 0:
            raise ValueError("ความยาวด้านต้องมากกว่า 0")
        self.__side = value

    # Property area (Read-only)
    @property
    def area(self):
        return self.__side * self.__side
s = Square(4)
print(f"Side: {s.side}, Area: {s.area}")  # Side: 4, Area: 16

s.side = 5
print(f"New Area: {s.area}")              # New Area: 25

#s.side = -10                              # Error: ความยาวด้านต้องมากกว่า 0

# s.area = 100                            # Error: AttributeError (Read-only)
```
