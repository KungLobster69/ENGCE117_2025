```python
def to_upper(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper
@to_upper
def greet():
    return "hello python"

print(greet())
```

```python
def add_brackets(func):
    def wrapper():
        result = func()
        return f"[{result}]"
    return wrapper
@add_brackets
def greet():
    return "hello python"

print(greet())
```

```python
@add_brackets
@to_upper
def greet():
    return "hello python"

print(greet())
```
