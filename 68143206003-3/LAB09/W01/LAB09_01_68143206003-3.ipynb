```python
class RPGCharacter :
    def __init__ (seft,name,hp,mp):
        seft.name = name
        seft.hp = hp
        seft.mp = mp
    def __str__(self):
        return f"{self.name} (HP:{self.hp},MP:{self.mp})"
    def __add__(seft,NewC):
        if isinstance(NewC, RPGCharacter):
            new_name = seft.name + NewC.name
            new_hp = seft.hp + NewC.hp
            new_mp = seft.mp + NewC.mp
            return RPGCharacter(new_name,new_hp,new_mp)
        elif isinstance(NewC,int):
            return RPGCharacter(seft.name,seft.hp + NewC,seft.mp)
        else :
            return NotImplemented
hero = RPGCharacter("Hero",100,50)
herobuf = hero + 50
print(herobuf)
```
