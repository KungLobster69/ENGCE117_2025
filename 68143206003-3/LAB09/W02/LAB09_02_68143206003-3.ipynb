```python
class Playlist :
    def __init__(self,song):
        self.song = list(song)
    def __len__(seft):
        return len(seft.song)
    def __getitem__(seft,index):
        if isinstance(index, slice):
            return Playlist(self.song[index])
        return seft.song[index]
    def __setitem__(self, index, value):
        self.song[index] = value
    def __contains__(self, item):
        return item in self.song
    p = Playlist(["A", "B", "C", "D"])

sub = p[0:2]

print(sub)         
print(sub[0])    
print(len(sub))  
```
