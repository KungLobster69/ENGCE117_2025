```python
class Square:
    def __init__(self,length):
        self.__side = length
    @property
    def side(self):
        return self.__side

    @side.setter
    def side(self,new_length):
        if new_length > 0 :
            self.__side = new_length
        else:
            print("Error: ความยาวด้านต้องมากกว่า 0")
    @property
    def area(self):
        return self.__side * self.__side

s = Square(4)
print(f"Side: {s.side}, Area: {s.area}") 

s.side = 5
print(f"New Area: {s.area}") 

s.side = -10
```
