```python
class Robot:
    def __init__(self, name, model):
        self.name = name
        self.model = model
      

r1 = Robot("Robot","R-01")
print(f"{r1.name}:{r1.model}")

```

```python
class Robot:

    def __init__(self, name, battery=100):
        self.name = name
        self.battery = battery
        
        if self.battery < 20:
            print(f"Warning: {self.name} has Low Battery!")


r3 = Robot("R-03", 15)
print(f"{r3.name} Battery:{r3.battery}%")
```
