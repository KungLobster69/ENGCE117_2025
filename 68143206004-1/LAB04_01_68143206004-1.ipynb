```python
def sum_rec(n):
    if n == 0 or n == 1:
      return n
    else:
       return n + sum_rec(n-1)
 

sum_rec(5)
```

```python
def reverse_str(s):
    if len(s) == 0 :
        return " "
    else:
        last = s[-1]     
        rest = s[:-1]    
        
        return last + reverse_str(rest)


reverse_str('ABC')
```

```python
def power(base, exp):
    if  exp == 0:
        return 1
    else:
        return base * power(base, exp - 1)
    
power(5,2)
```
