```python
def even_gen(limit):
    for num in range(limit + 1):
        if num % 2 == 0:
            yield num  


for n in even_gen(10):
    print(n)
```

```python
def frange(start, stop, step):

    current = start
    
    while current < stop:
        yield current    
        current += step   

for n in frange(0, 1, 0.5):
    print(n)

print("" * 20)

for n in frange(0.5, 2.5, 0.5):
    print(n)
```

```python
def number_gen(set):
    for i in range(1, set + 1):
        yield i

def square_gen(numbers):
    for n in numbers:
        yield n ** 2


for result in square_gen(number_gen(5)):
    print(result)
```
