```python
class Robot:
    count = 0
    category = "Mechanical"

    def __init__(self, name):
      self.name = name
      Robot.count +=1

r1 = Robot("Motor")

print(f"Part:{r1.name}, Type:{r1.category}")
```

```python
class Robot:
    total_count = 0
    
    def __init__(self, name):
      self.name = name
      Robot.total_count +=1

r1 = Robot("R1")
r2 = Robot("R2")
r3 = Robot("R3")

print(f"total parts in system:{Robot.total_count}")
```

```python

p1 = Robot("Senser") 

is_equal = p1.total_count == Robot.total_count

print(f"p1.total_count: {p1.total_count}")
print(f"Robot.total_count: {Robot.total_count}")
print(f"Are they the same? {is_equal}")
```
