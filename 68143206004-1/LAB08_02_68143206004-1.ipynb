```python
class WiFi:
    def connect(self):
        print("📡 Connected to WiFi network")

class Bluetooth:
    def connect(self):
        print("🔵 Paired with Bluetooth device")

W = WiFi()
B = Bluetooth()

W.connect()
B.connect()


def initialize_system(device):
        print("System initializing...")
        device.connect()

initialize_system(W)
initialize_system(B)
```

```python
def initialize_system(device):
    try:
        print("System initializing...")
        device.connect()
    except AttributeError:
        print(f"Error: Object '{device}' ไม่มีฟังก์ชัน connect")

initialize_system(W)
initialize_system(B)
initialize_system(1234)
initialize_system("TOT")
initialize_system("3BB")
```
