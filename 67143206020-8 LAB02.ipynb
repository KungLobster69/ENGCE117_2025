```python
def get_stats(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    return min_val, max_val  


nums = [1, 2, 3, 4, 5]
min_val, max_val = get_stats(nums)   

print(f"Min: {min_val}, Max: {max_val}")
```

```python
from collections import Counter

text = "apple banana apple orange banana apple"

# แยกคำเป็นรายการ
words = text.split()

# นับจำนวนคำ
count = Counter(words)

print("จำนวนคำแต่ละคำ:", count)

# หาคำที่ซ้ำบ่อยที่สุด 1 คำ
most_common_word = count.most_common(1)

print("คำที่ซ้ำบ่อยที่สุด:", most_common_word)
```

```python
from collections import defaultdict

names = ["Alice", "Bob", "Charlie", "David", "Eve"]

# defaultdict แบบ list
grouped = defaultdict(list)

# จัดกลุ่มตามความยาว
for name in names:
    grouped[len(name)].append(name)

# แสดงผล
for length, group in grouped.items():
    print(length, ":", group)
```

```python
from collections import deque

# สร้าง deque ที่เก็บได้สูงสุด 3 รายการ
logs = deque(maxlen=3)

# รับข้อความเข้า 5 ตัว
inputs = ["Log1", "Log2", "Log3", "Log4", "Log5"]

for msg in inputs:
    logs.append(msg)
    print("เพิ่ม:", msg, "=>", list(logs))

print("\nLog ที่เหลือสุดท้าย:", list(logs))
```
