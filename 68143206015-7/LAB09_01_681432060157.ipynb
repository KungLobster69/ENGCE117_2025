```python
#Level 2
class RPGCharacter:
    def __init__(self, name, hp, mp):
        self.name = name
        self.hp = hp
        self.mp = mp
    def __str__(self):
        return f"{self.name} (HP: {self.hp}, MP: {self.mp})"
    def __add__(self, other):
        new_name = self.name + "-" + other.name
        new_hp = self.hp + other.hp
        new_mp = self.mp + other.mp
        return RPGCharacter(new_name, new_hp, new_mp)
c1 = RPGCharacter("Hero", 100, 50)
c2 = RPGCharacter("Mage", 80, 120)
c3 = c1 + c2
print(c3)
```

```python
#Level 3
class RPGCharacter:
    def __init__(self, name, hp, mp):
        self.name = name
        self.hp = hp
        self.mp = mp
    def __str__(self):
        return f"{self.name} (HP: {self.hp}, MP: {self.mp})"
    def __add__(self, other):
        if isinstance(other, RPGCharacter):
            return RPGCharacter(
                self.name + "-" + other.name,
                self.hp + other.hp,
                self.mp + other.mp
            )
        elif isinstance(other, int):
            return RPGCharacter(
                self.name,
                self.hp + other,
                self.mp
            )
c1 = RPGCharacter("Hero", 100, 50)
heal = c1 + 50
print(heal)
```
