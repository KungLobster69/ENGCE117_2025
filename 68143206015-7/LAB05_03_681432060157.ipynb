```python
class RobotPart:
    total_count = 0   # ตัวแปรระดับคลาส (shared)

    def __init__(self, name):
        self.name = name
        RobotPart.total_count += 1


# สร้างชิ้นส่วนเดิม
p1 = RobotPart("Arm")

# สร้างชิ้นส่วนเพิ่มอีก 1 ชิ้น
p2 = RobotPart("Leg")

# ตรวจสอบว่า total_count ผ่าน object กับผ่าน class เป็นค่าเดียวกันหรือไม่
print("Is it shared?", p1.total_count == RobotPart.total_count)
```
