```python
class Robot:
    def __init__(self, name, model):
        self.name = name
        self.model = model

    def __str__(self):
        return f"Robot: {self.model} ({self.name})"


r1 = Robot("Exception", "R-01")
print(r1)
```

```python
class Robot:
    def __init__(self, name, model, battery = 100):
        self.name = name
        self.model = model
        self.battery = battery

        
    def __str__(self):
        return f"Robot: {self.model} ({self.name}) - Battery {self.battery}%"


r1 = Robot("Exception", "R-02")
print(r1)
```
