```python
def flexi_cals(mode, *nums):
    if mode == "add":
        total = 0
        for item in nums:
            total = total + item
        return total

    elif mode == "mult":
        total = 1
        for item in nums:
            total *= item
        return total

    else:
        return "Mode Error"
```

```python
total = flexi_cals("add",35,55)
print (total)
```
