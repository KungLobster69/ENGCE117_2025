```python
class Character:
    def __init__(self, name, hp, power):
        self.name = name
        self.hp = hp
        self.power = power

    def take_damage(self, amount):
        self.hp -= amount
        print(f"{self.name} takes {amount} damage! HP left: {self.hp}")

    def attack(self, target):
        print(f"{self.name} attacks {target.name}!")
        target.take_damage(self.power)

    def is_alive(self):
        return self.hp > 0


class Warrior(Character):
    def __init__(self, name, hp, power, defense):
        super().__init__(name, hp, power)
        self.defense = defense

    def take_damage(self, amount):
        reduced_damage = max(amount - self.defense, 0)
        self.hp -= reduced_damage
        print(
            f"{self.name} blocks with defense! "
            f"Damage taken: {reduced_damage}, HP left: {self.hp}"
        )


class Mage(Character):
    def __init__(self, name, hp, power, mana):
        super().__init__(name, hp, power)
        self.mana = mana

    def attack(self, target):
        if self.mana >= 10:
            self.mana -= 10
            damage = self.power + 10
            print(
                f"{self.name} casts a spell on {target.name}! "
                f"Mana left: {self.mana}"
            )
            target.take_damage(damage)
        else:
            print(f"{self.name} has not enough mana!")
            super().attack(target)
```

```python
class Character:
    def __init__(self, name, hp, power):
        self.name = name
        self.hp = hp
        self.power = power

    def take_damage(self, amount):
        self.hp -= amount
        print(f"{self.name} takes {amount} damage! HP left: {self.hp}")

    def attack(self, target):
        print(f"{self.name} attacks {target.name}")
        target.take_damage(self.power)

    def is_alive(self):
        return self.hp > 0


class Warrior(Character):
    def __init__(self, name, hp, power, defense):
        super().__init__(name, hp, power)
        self.defense = defense

    def take_damage(self, amount):
        damage = max(amount - self.defense, 0)
        self.hp -= damage
        print(
            f"{self.name} blocks! Damage taken: {damage}, HP left: {self.hp}"
        )


class Mage(Character):
    def __init__(self, name, hp, power, mana):
        super().__init__(name, hp, power)
        self.mana = mana

    def attack(self, target):
        if self.mana >= 10:
            self.mana -= 10
            damage = self.power + 10
            print(
                f"{self.name} casts spell on {target.name}! "
                f"Mana left: {self.mana}"
            )
            target.take_damage(damage)
        else:
            print(f"{self.name} has no mana, normal attack!")
            super().attack(target)


# ===== Battle Simulation =====
warrior = Warrior("Thor", 120, 15, defense=5)
mage = Mage("Merlin", 80, 10, mana=40)

attacker = warrior
defender = mage
turn = 1

while warrior.is_alive() and mage.is_alive():
    print(f"\n--- Turn {turn} ---")
    attacker.attack(defender)

    # ‡∏ñ‡πâ‡∏≤‡∏ù‡πà‡∏≤‡∏¢‡∏£‡∏±‡∏ö‡∏ï‡∏≤‡∏¢ ‡∏à‡∏ö‡πÄ‡∏Å‡∏°
    if not defender.is_alive():
        break

    # ‡∏™‡∏•‡∏±‡∏ö‡∏Ñ‡∏ô‡πÇ‡∏à‡∏°‡∏ï‡∏µ
    attacker, defender = defender, attacker
    turn += 1

print("\n===== Battle Result =====")
print(f"{attacker.name} wins! üèÜ")
```
