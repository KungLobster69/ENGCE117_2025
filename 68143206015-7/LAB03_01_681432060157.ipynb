```python
def flexi_cals(mode, *nums):
    if mode == "add":
        total = 0
        for item in nums:
            total = total + item
        return total

    elif mode == "mult":
        total = 1
        for item in nums:
            total *= item
        return total

    else:
        return "Mode Error"
```

```python
total = flexi_cals("add",35,55)
print (total)
```

```python
def stat_calc(mode, *args):
    if len(args) == 0:
        return 0
    if mode == "avg":
        return sum(args) / len(args)
    elif mode == "max":
        return max(args)
print(stat_calc("avg", 10, 20, 30))  
print(stat_calc("avg")) 
```
