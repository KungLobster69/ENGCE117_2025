```python
class WiFi:
    def connect(self):
        print("Connected to WiFi network")

class Bluetooth:
    def connect(self):
        print("Paired with Bluetooth device")

wifi = WiFi()
bluetooth = Bluetooth()

wifi.connect()
bluetooth.connect()
```

```python
def initialize_system(device):
    print("System initializing...")
    device.connect()

wifi = WiFi()
bluetooth = Bluetooth()

initialize_system(wifi)
initialize_system(bluetooth)
```

```python
def initialize_safe(device):
    print("System initializing...")
    try:
        device.connect()
    except AttributeError:
        print("Error:เชื่อมต่อไม่ได้")
```

```python
initialize_safe(wifi)
initialize_safe(bluetooth)
```

```python
initialize_safe("1")
initialize_safe(1)
```
