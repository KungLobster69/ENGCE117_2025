```python
from abc import ABC, abstractmethod

class IoTDevice(ABC):
    @abstractmethod
    def turn_on(self):
        pass

    @abstractmethod
    def turn_off(self):
        pass

class SmartLight(IoTDevice):
    def turn_on(self):
        print("ไฟเปิด")

    def turn_off(self):
        print("ไฟปิด")

light = SmartLight()

light.turn_on()
light.turn_off()
```

```python
from abc import ABC, abstractmethod

class IoTDevice(ABC):
    @abstractmethod
    def turn_on(self):
        pass

    @abstractmethod
    def turn_off(self):
        pass

class SmartSpeaker(IoTDevice):
    def turn_on(self):
        print("ลำโพงเปิด")
    
    def turn_off(self):
        print("ลำโพงปิด")

speaker = SmartSpeaker()

speaker.turn_on()
speaker.turn_off()
```

```python
def test_device(device):
    if isinstance(device, IoTDevice):
        print(f"Testing {device.__class__.__name__}...")
        device.turn_on()
        device.turn_off()
    else:
        print("ไม่ใช่อุปกรณ์ IoT ที่รองรับ")

light = SmartLight()

test_device(light)

test_device("ไม่ใช่อุปกรณ์ IoT")
```
