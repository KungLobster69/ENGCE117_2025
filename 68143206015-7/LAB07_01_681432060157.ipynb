```python
class Animal:
    def make_sound(self):
        print("... (Silence)")


class Dog(Animal):
    def make_sound(self):
        print("Woof! Woof!")


class Cat(Animal):
    def make_sound(self):
        print("Meow~ Meow~")


# List ‡∏Ç‡∏≠‡∏á‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å‡πÉ‡∏ô‡∏ß‡∏á‡∏î‡∏ô‡∏ï‡∏£‡∏µ üéµ
band = [Dog(), Cat(), Dog(), Animal()]

# ‡∏ß‡∏ô‡∏•‡∏π‡∏õ‡πÉ‡∏´‡πâ‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡∏£‡πâ‡∏≠‡∏á‡πÄ‡∏û‡∏•‡∏á
for member in band:
    member.make_sound()
```
