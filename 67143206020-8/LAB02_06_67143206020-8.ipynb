```python
from collections import Counter

text = "apple banana apple orange banana apple"

words = text.split()

count = Counter(words)

print("จำนวนคำแต่ละคำ:", count)


most_common_word = count.most_common(1)

print("คำที่ซ้ำบ่อยที่สุด:", most_common_word)
```

```python
from collections import defaultdict

names = ["Alice", "Bob", "Charlie", "David", "Eve"]

grouped = defaultdict(list)


for name in names:
    grouped[len(name)].append(name)


for length, group in grouped.items():
    print(length, ":", group)
```

```python
from collections import deque


logs = deque(maxlen=3)

inputs = ["Log1", "Log2", "Log3", "Log4", "Log5"]

for msg in inputs:
    logs.append(msg)
    print("เพิ่ม:", msg, "=>", list(logs))
```
