```python
class Square:
    def __init__(self, side):
        self.__side = side

    def set_side(self, s):
        if s > 0:
            self.__side = s
        else:
            print("ค่าความยาวต้องมากกว่า 0")

    def get_side(self):
        return self.__side

    def area(self):
        return self.__side ** 2


sq = Square(5)

print(sq.get_side())
print(sq.area())

sq.set_side(10)
print(sq.area())

sq.set_side(-3)
```
