```python
def sum_rec(n):
    if n == 1:
        return 1
    return n + sum_rec(n-1)
print (sum_rec(4))
```

```python
def reverse_str(s):
    if s  ==  "":
        return ""
    return reverse_str(s[1:]) + s[0]
print (reverse_str("ABC"))
    
```

```python
def power(a,b):
    if b == 0:
        return 1
    return a * power(a,b-1)
print(power(2,4))
```
