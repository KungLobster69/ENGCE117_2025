```python

class Character:
    def __init__(self, name, hp, power):
        self.name = name
        self.hp = hp
        self.power = power

    def take_damage(self, amount):
        self.hp -= amount
        print(f"{self.name} ได้รับความเสียหาย {amount} HP เหลือ {self.hp} HP")

    def attack(self, target):
        print(f"{self.name} โจมตี {target.name}")
        target.take_damage(self.power)

    def is_alive(self):
        return self.hp > 0


class Warrior(Character):
    def __init__(self, name, hp, power, defense):
        super().__init__(name, hp, power)
        self.defense = defense

    def take_damage(self, amount):
        damage = amount - self.defense
        if damage < 0:
            damage = 0
        self.hp -= damage
        print(
            f"{self.name} ป้องกันได้ {self.defense} "
            f"ได้รับความเสียหาย {damage} HP เหลือ {self.hp} HP"
        )


class Mage(Character):
    def __init__(self, name, hp, power, mana):
        super().__init__(name, hp, power)
        self.mana = mana

    def attack(self, target):
        if self.mana >= 10:
            self.mana -= 10
            damage = self.power + 15
            print(
                f"{self.name} ใช้เวทโจมตี {target.name} "
                f"(ใช้ mana 10 เหลือ {self.mana})"
            )
            target.take_damage(damage)
        else:
            print(f"{self.name} mana ไม่พอ ใช้โจมตีปกติ")
            super().attack(target)

warrior = Warrior("Knight", 120, 20, 5)
mage = Mage("Wizard", 80, 15, 30)

mage.attack(warrior)
warrior.attack(mage)
```

```python

warrior = Warrior("Knight", 120, 20, 5)
mage = Mage("Wizard", 80, 15, 30)

round_num = 1
while warrior.is_alive() and mage.is_alive():
    print(f"\n--- รอบที่ {round_num} ---")

  
    warrior.attack(mage)
    if not mage.is_alive():
        print(f"\n{warrior.name} เป็นผู้ชนะ!")
        break

   
    mage.attack(warrior)
    if not warrior.is_alive():
        print(f"\n{mage.name} เป็นผู้ชนะ!")
        break

    round_num += 1
```
