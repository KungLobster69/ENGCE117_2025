```python
def to_upper(func):
    def wrapper():
        result = func()         
        return result.upper()   
    return wrapper


@to_upper
def greeting():
    return "holle world"


print(greeting())
```

```python
def to_upper(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper


def add_brackets(func):
    def wrapper():
        result = func()
        return f"[{result}]"
    return wrapper


@add_brackets
@to_upper
def greeting():
    return "hello world"


print(greeting())
```
