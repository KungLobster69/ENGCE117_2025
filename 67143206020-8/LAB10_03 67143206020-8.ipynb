```python
import tkinter as tk

root = tk.Tk()


username_var = tk.StringVar()
accept_var = tk.BooleanVar()


label = tk.Label(root, text="Username:")
label.pack()


entry = tk.Entry(root, textvariable=username_var)
entry.pack()

checkbox = tk.Checkbutton(root, text="I accept the terms", variable=accept_var)
checkbox.pack()

root.mainloop()
```

```python
import tkinter as tk

root = tk.Tk()

username_var = tk.StringVar()
accept_var = tk.BooleanVar()

tk.Label(root, text="Username:").pack()
tk.Entry(root, textvariable=username_var).pack()
tk.Checkbutton(root, text="I accept the terms", variable=accept_var).pack()
def submit():
    username = username_var.get()
    accepted = accept_var.get()

    if accepted:
        print("Username:", username)
    else:
        print("กรุณายอมรับเงื่อนไขก่อน!")


tk.Button(root, text="Submit", command=submit).pack()

root.mainloop()
```
