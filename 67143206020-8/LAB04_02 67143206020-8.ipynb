```python
def even_gen(limit):
    for i in range(limit + 1):
        if i % 5 == 0:
            yield i
for n in even_gen(20):
    print(n)
```

```python
def frange(start, stop, step):
    x = start
    while x < stop:
        yield x
        x += step

for n in frange(0, 1, 0.5):
    print(n)
```

```python
# Gen1: ผลิตเลข 1–5
def gen_numbers():
    for i in range(1, 6):
        print(f"Gen1 ผลิต: {i}")
        yield i


# Gen2: รับเลข -> ยกกำลังสอง -> ส่งต่อ
def gen_square(numbers):
    for n in numbers:
        print(f"Gen2 แปรรูป: {n} -> {n*n}")
        yield n * n


# ต่อท่อ (Pipeline)
pipeline = gen_square(gen_numbers())

for result in pipeline:
    print(f"ผลลัพธ์สุดท้าย: {result}")
```
