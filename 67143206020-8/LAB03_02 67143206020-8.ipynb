```python
raw_names = [" somchai ","alice"," BOB","c"]

c = list(map(lambda w: w.strip().lower(), raw_names))
n = filter(lambda name: len(name) > 2, c)

print(list(n))
```

```python
raw_names = [" somchai ","alice"," BOB","c"]
result = [w.strip().lower() for w in raw_names if len(w.strip()) > 2]
print(result)
```
