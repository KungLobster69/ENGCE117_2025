```python
def super_sum(*nums):
    return sum(nums)
print(super_sum(1, 2, 3))        
print(super_sum(10, 20, 30, 40)) 
```

```python
def flexi_calc(mode, *args):
    if mode == "add":
        return sum(args)
    elif mode == "mult":
        result = 1
        for num in args:
            result *= num
        return result
    else:
        return "Error: mode ต้องเป็น 'add' หรือ 'mult' เท่านั้น"
print(flexi_calc("add", 1, 2, 3, 4))     
print(flexi_calc("mult", 2, 3, 4))       
print(flexi_calc("x", 1, 2))             
```

```python
def stat_calc(mode, *args):
    if not args:
        return 0

    if mode == "avg":
        return sum(args) / len(args)
    elif mode == "max":
        return max(args)
    else:
        return "Error: mode ต้องเป็น 'avg' หรือ 'max'"
print(stat_calc("avg", 10, 20, 30))  
print(stat_calc("avg"))              
print(stat_calc("max", 5, 9, 2))     
print(stat_calc("max"))             
```
