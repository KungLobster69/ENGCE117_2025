```python
class Circle:
    def draw(self):
        print(" O วาดวงกลม")
 


class Square:
    def draw(self):
        print(" # วาดสี่เหลี่ยม") 


def render_screen(shape_list):
    for shapes in shape_list:
        shapes.draw()

class Triangle:
    def draw(self):
        print(" ^ วาดสามเหลี่ยม")

shapes = [
    Circle(),
    Square(),
    Triangle(),
    Circle(), 
    Square()
    ]
render_screen(shapes)                
```
