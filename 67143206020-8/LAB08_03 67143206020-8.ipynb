```python
from abc import ABC, abstractmethod

class IoTDevice(ABC):

    @abstractmethod
    def turn_on(self):
        pass

    @abstractmethod
    def turn_off(self):
        pass


class SmartLight(IoTDevice):

    def turn_on(self):
        print(" ไฟเปิด")

    def turn_off(self):
        print(" ไฟปิด")


def test_device(device):
    if isinstance(device, IoTDevice):
        print("Testing SmartLight...")
        device.turn_on()
        device.turn_off()
    else:
        print("ไม่ใช่อุปกรณ์ IoT ที่รองรับ")



test_device(SmartLight())      
test_device("หลอดไฟปลอม")     
```
