```python
def even_gen(limit):
    for i in range(limit + 1):
        if i % 2 == 0:
            yield i
for num in even_gen(10):
    print(num)
```

```python
def frange(start, stop, step):
    while start < stop:
        yield round(start, 10)  
        start += step
for num in frange(0, 1, 0.5):
    print(num)

```

```python
def gen1():
    for i in range(1, 6):
        yield i

def gen2(input_gen):
    for num in input_gen:
        yield num ** 2
for result in gen2(gen1()):
    print(result)
```
