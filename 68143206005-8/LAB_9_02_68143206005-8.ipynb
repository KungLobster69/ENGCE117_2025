```python
class Playlist:
    def __init__(self, songs):
        self.songs = list(songs)   # เก็บเป็น list ของ string

    # Level 1
    def __len__(self):
        return len(self.songs)

    def __getitem__(self, index):
        # Level 3: รองรับ slicing
        if isinstance(index, slice):
            return Playlist(self.songs[index])  # ต้องคืนเป็น Playlist ใหม่
        return self.songs[index]

    # Level 2
    def __setitem__(self, index, value):
        self.songs[index] = value
    def __contains__(self, item):
        return item in self.songs
    def __str__(self):
        return f"Playlist: {self.songs}"
    
    p = Playlist(["Song Seven ", "Song Wannabe ", "Song Perfect"])

# Level 1
    print(len(p))        # 3
    print(p[0])          # Song A

# Level 2
    p[0] = "New Song"
    print(p[0])          # New Song
    print("Song B" in p) # True

# Level 3
    p2 = p[0:2]
    print(p2)            # Playlist: ['New Song', 'Song B']
    print(type(p2))      # <class '__main__.Playlist'>

```
