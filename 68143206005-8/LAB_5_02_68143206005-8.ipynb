```python
class Robot:
    def __init__(self, name, model):
        self.name = name
        self.model = model

r1 = Robot("TaTar", "Explorer")
print(f"Robot: {r1.name} ({r1.model})")
```

```python
class Robot:
    def __init__(self, name, model, battery=100):
        self.name = name     
        self.model = model      
        self.battery = battery 
r2 = Robot("TaTaya", "Worker")
print(f"{r2.name} Battery: {r2.battery}%")
```
