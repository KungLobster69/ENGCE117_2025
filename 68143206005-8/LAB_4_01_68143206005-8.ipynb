```python
def sum_rec(n):
    if n == 0: return 0
    if n == 1: return 1
    else:
        return n + sum_rec(n - 1)
print(sum_rec(5))  
 
```

```python
def reverse_str(s):
    if len(s) <= 1:
        return s
    else:
        return s[-1] + reverse_str(s[:-1])
print(reverse_str("ABC"))  

```

```python
def power(base, exp):
    if exp == 0:
        return 1
    else:
        return base * power(base, exp - 1)

print(power(2, 3))

```
