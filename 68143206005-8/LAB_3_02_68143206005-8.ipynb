```python
raw_names = [" somchai ", "alice", " BOB", "c"]

map_name = list(map(lambda name: name.strip().lower(), raw_names))


print(map_name)
```

```python
filter_name =  list(filter(lambda name:  len(name)>=2, map_name))
print(filter_name)
```

```python
res_comp = [name.strip().lower() for name in raw_names if  len(name)>=2]
print(res_comp)
```
