```python
class RPGCharacter:
    def __init__(self, name, hp, mp):
        self.name = name
        self.hp = hp
        self.mp = mp

   
    def __str__(self):
        return f"{self.name} (HP: {self.hp}, MP: {self.mp})"

 # Level 2 และ 3
    def __add__(self, other):
        # Level 2: บวกตัวละครกับตัวละคร
        if isinstance(other, RPGCharacter):
            new_name = self.name + other.name
            new_hp = self.hp + other.hp
            new_mp = self.mp + other.mp
            return RPGCharacter(new_name, new_hp, new_mp)

        # Level 3: บวกกับตัวเลข (เพิ่ม hp อย่างเดียว)
        elif isinstance(other, int):
            return RPGCharacter(self.name, self.hp + other, self.mp)

        else:
            return NotImplementedError
        
        
char1 = RPGCharacter("Pro", 100, 50)
char2 = RPGCharacter("Be", 60, 120)

print(char1)  
# pro (HP: 100, MP: 50)

# Level 2
char3 = char1 + char2
print(char3)  
# pb (HP: 160, MP: 170)

# Level 3
char4 = char1 + 50
print(char4)  
# pr (HP: 150, MP: 50)


```
