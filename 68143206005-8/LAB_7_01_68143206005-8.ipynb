```python
class Animal:
    def make_sound(self):
        print("...(Silence)")
```

```python
class Dog(Animal):
    def make_sound(self):
        print("Woof! Woof!")
class Cat(Animal):
    def make_sound(self):
        print("Meow~")
```

```python
zoo =  [Dog(), Cat(), Dog(), Animal()]
print("---zoo band---")
for animal in zoo:
    animal.make_sound()
    print("Meow~")
```
