```python
names = ["alice", "bob", "charlie"]
names_upper = [name.upper() for name in names]
print(names_upper)  
# ['ALICE', 'BOB', 'CHARLIE']
```

```python
prices = [100, 250, 80, 500, 120]
vat_prices = [price * 1.07 for price in prices if price > 200]
print(vat_prices)
# [267.5, 535.0]
```

```python
currencies = {'USD': 35, 'EUR': 38, 'JPY': 0.25}
inverted = {value: key for key, value in currencies.items() if value > 1}
print(inverted)
# {35: 'USD', 38: 'EUR'}
```
