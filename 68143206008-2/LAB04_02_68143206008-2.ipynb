```python
def even_gen(limit):
    for i in range(0, limit + 1):
        if i % 2 == 0:
            yield i

for x in even_gen(10):
    print(x)
```

```python
def frange(start, stop, step):
    current = start
    while current < stop:
        yield current
        current += step

for x in frange(0, 1, 0.5):
    print(x)
```

```python
def gen_numbers():
    for i in range(1, 6):
        yield i

def square_gen(source):
    for x in source:
        yield x * x

pipeline = square_gen(gen_numbers())

for value in pipeline:
    print(value)
```
