```python
def to_upper(func):
    def wrapper():
        text = func()
        return text.upper()
    return wrapper

@to_upper
def greet():
    return "hello python"

print(greet())
```

```python
def add_brackets(func):
    def wrapper():
        text = func()
        return f"[{text}]"
    return wrapper

@add_brackets
def greet():
    return "hello python"

print(greet())
```

```python
@add_brackets
@to_upper
def greet():
    return "hello python"

print(greet())
```
