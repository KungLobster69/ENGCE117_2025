```python
class Animal:
    def make_sound(self):
        print("... (Silence)")
```

```python
class Dog(Animal):
    def make_sound(self):
        print("Woof! Woof!")

class Cat(Animal):
    def make_sound(self):
        print("Meow~ Meow~")
```

```python
band = [Dog(), Cat(), Dog(), Animal()]

for member in band:
    member.make_sound()
```
