```python
# เพิ่ม logic ตรวจสอบแบตเตอรี่ต่ำใน __init__
class Robot:
    def __init__(self, name, model, battery=100):
        self.name = name
        self.model = model
        self.battery = battery

        # ตรวจสอบว่าแบตเตอรี่ต่ำกว่า 20 หรือไม่
        if self.battery < 20:
            print(f"Warning: {self.name} has Low Battery!")

# สร้าง object ที่มีแบตเตอรี่ต่ำ
r3 = Robot("R-03", "Scout", 19)
```
