```python
import tkinter as tk

window = tk.Tk()
window.title("Workshop 2")
window.geometry("200x150")  

label = tk.Label(window, text="ยังไม่ได้คลิก")
label.pack(pady=20)

def change_text():
    label.config(text="คลิกแล้วจ้า!", fg="red")

btn = tk.Button(window, text="คลิกฉันสิ!", command=change_text)
btn.pack()

window.mainloop()
```
