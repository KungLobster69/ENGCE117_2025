```python
def get_stats(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    return (min_val, max_val)

nums = [1, 2, 3, 4, 5]

min_val, max_val = get_stats(nums)

print(f"Min: {min_val}, Max: {max_val}")
```
