```python
import tkinter as tk

window = tk.Tk()
window.title("Numpad")
window.geometry("300x350")

entry = tk.Entry(window, font=("Arial", 16))
entry.grid(row=0, column=0, columnspan=3, padx=10, pady=10, sticky="we")

numbers = 1
for r in range(1, 4):    
    for c in range(3):     
        btn = tk.Button(window, text=str(numbers), width=8, height=2)
        btn.grid(row=r, column=c, padx=5, pady=5)
        numbers += 1

btn0 = tk.Button(window, text="0", width=20, height=2)
btn0.grid(row=4, column=0, columnspan=2, padx=5, pady=5)

btn_dot = tk.Button(window, text=".", width=8, height=2)
btn_dot.grid(row=4, column=2, padx=5, pady=5)

window.mainloop()
```
