```python
inventory = {'apple': 10, 'banana': 20}

def check_price(item_name):
    return inventory.get(item_name, 0)

print("Check Durian:", check_price("durian")) 
# print("Check Apple:", check_price("apple"))     
```

```python
inventory = {'apple': 10, 'banana': 20}

def add_stock(item, quantity):
    inventory[item] = inventory.get(item, 0) + quantity
    return inventory

print(add_stock("apple", 5))
# print(add_stock("orange", 12)) 
```

```python
inventory = {'apple': 15, 'banana': 20}
branch2 = {'banana': 5, 'orange': 10}

def merge_inventory(main, branch):
    for item, qty in branch.items():
        main[item] = main.get(item, 0) + qty
    return main

print(merge_inventory(inventory, branch2))
```
