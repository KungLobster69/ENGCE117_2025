```python
class Shape:
    def draw(self):
        pass

class Circle(Shape):
    def draw(self):
        print("ğŸ”´ à¸§à¸²à¸”à¸§à¸‡à¸à¸¥à¸¡")

class Square(Shape):
    def draw(self):
        print("ğŸŸ¦ à¸§à¸²à¸”à¸ªà¸µà¹ˆà¹€à¸«à¸¥à¸µà¹ˆà¸¢à¸¡")

class Triangle(Shape):
    def draw(self):
        print("ğŸ”º à¸§à¸²à¸”à¸ªà¸²à¸¡à¹€à¸«à¸¥à¸µà¹ˆà¸¢à¸¡")


def render_screen(shape_list):
    print("--- Rendering 3 shapes ---")
    for s in shape_list:
        s.draw()


shapes = [Circle(), Triangle(), Square()]
render_screen(shapes)
```
