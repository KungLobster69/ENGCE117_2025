```python
# สร้างคลาส RobotPart
class RobotPart:
    # class attribute ใช้ร่วมกันทุก object
    category = "Mechanical"

    # __init__ รับชื่อชิ้นส่วน
    def __init__(self, name):
        self.name = name  # instance attribute เป็นของแต่ละ object

# สร้างชิ้นส่วน 1 ชิ้น
p1 = RobotPart("Motor")

# แสดงชื่อชิ้นส่วนและประเภท
print(f"Part: {p1.name}, Type: {p1.category}")

```

```python
class RobotPart:
    # class attribute ใช้นับจำนวนชิ้นส่วนทั้งหมด
    category = "Mechanical"
    total_count = 0  # เริ่มต้นที่ 0

    def __init__(self, name):
        self.name = name          # ชื่อชิ้นส่วน
        RobotPart.total_count += 1  # เพิ่มจำนวนทุกครั้งที่สร้าง object ใหม่

# สร้างชิ้นส่วน 2 ชิ้น
p1 = RobotPart("Motor")
p2 = RobotPart("Sensor")

# แสดงจำนวนชิ้นส่วนทั้งหมดในระบบ
print("Total parts in system:", RobotPart.total_count)
```

```python
# สร้างชิ้นส่วนเพิ่มอีก 1 ชิ้น
p3 = RobotPart("Camera")

# ตรวจสอบว่าค่า total_count ของ object และของคลาสเป็นค่าเดียวกันหรือไม่
print("Is it shared?", p3.total_count is RobotPart.total_count)
```
