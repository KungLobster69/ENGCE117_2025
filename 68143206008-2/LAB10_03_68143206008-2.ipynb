```python
import tkinter as tk

window = tk.Tk()
window.title("Registration Form")
window.geometry("300x200")

username_var = tk.StringVar()
terms_var = tk.BooleanVar()

label_user = tk.Label(window, text="Username:")
label_user.pack(pady=5)

entry_user = tk.Entry(window, textvariable=username_var)
entry_user.pack()

check_terms = tk.Checkbutton(window, text="I accept the terms", variable=terms_var)
check_terms.pack(pady=10)

def submit():
    username = username_var.get()
    accepted = terms_var.get()

    if username == "":
        print("กรุณากรอกชื่อผู้ใช้")
        return
    
    if not accepted:
        print("กรุณายอมรับเงื่อนไขก่อน")
        return
    
    print("สมัครสมาชิกสำเร็จ!")
    print("Username:", username)

btn_submit = tk.Button(window, text="Submit", command=submit)
btn_submit.pack()

window.mainloop()
```
