```python
class Robot:
    pass  

r1 = Robot()
r2 = Robot()

print(r1)
print(r2)

```

```python
r1.name = "Commander"

r1.battery = 100

print(f"Robot Name: {r1.name}, Battery: {r1.battery}%")
```

```python
r2.name = "Commander"

r2.battery = 100

print(f"Robot Name: {r2.name}, Battery: {r2.battery}%")
```

```python
# ให้ r2 อ้างอิงไปยัง object เดียวกับ r1 (ไม่ใช่การสร้าง object ใหม่)
# r2 = r1

# ตรวจสอบว่า r1 และ r2 อ้างอิง object เดียวกันหรือไม่
print("Is r1 the same as r2?", r1 is r2)

# แสดงที่อยู่ของ r1 ในหน่วยความจำ (RAM)
print("Address r1:", id(r1))

# แสดงที่อยู่ของ r2 ในหน่วยความจำ (RAM)
print("Address r2:", id(r2))
```
