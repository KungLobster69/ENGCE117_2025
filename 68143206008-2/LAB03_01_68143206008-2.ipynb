```python
def super_sum(*nums):
    # *nums คือการรับอาร์กิวเมนต์จำนวนเท่าไรก็ได้ และเก็บเป็น tuple
    return sum(nums)  # sum() จะนำค่าทั้งหมดใน nums มาบวกรวมกัน

# เรียกฟังก์ชันพร้อมส่งตัวเลข 3 ตัว ค่าใน nums = (10, 20, 30)
print(super_sum(10, 20, 30))  # ผลลัพธ์: 60

# เรียกฟังก์ชันโดยไม่ส่งอาร์กิวเมนต์ ค่าใน nums = ()
print(super_sum())  # ผลลัพธ์: 0 (sum ของ tuple ว่างคือ 0)
```

```python
def flexi_calc(mode, *args):
    # mode คือโหมดการคำนวณ เช่น "add" หรือ "mult"
    # *args คืออาร์กิวเมนต์จำนวนเท่าไรก็ได้ (ตัวเลขหลายตัว)

    if mode == "add":
        # โหมดบวก: ใช้ sum() รวมค่าทั้งหมดใน args
        return sum(args)

    elif mode == "mult":
        # โหมดคูณ: เริ่มต้นด้วย 1 แล้วนำทุกตัวใน args มาคูณสะสม
        result = 1
        for n in args:
            result *= n
        return result

    else:
        # ถ้า mode ไม่ถูกต้อง จะคืนค่า None
        return None   
    
# เรียกโหมดบวก: 10 + 20 + 30 = 60
print(flexi_calc("add", 10, 20, 30))  

# เรียกโหมดคูณ: 2 × 3 × 4 = 24
print(flexi_calc("mult", 2, 3, 4))  

# เรียกโหมดบวกแต่ไม่ส่งตัวเลข args = ()
# sum(()) = 0 ดังนั้นผลลัพธ์ = 0
# print(flexi_calc("add"))  
              

```

```python
def stat_calc(mode, *args):
    # mode คือโหมดการคำนวณ เช่น "avg" หรือ "max"
    # *args คือค่าตัวเลขจำนวนเท่าไรก็ได้

    # ถ้าไม่มีการส่งตัวเลขเข้ามาเลย (args ว่าง)
    if len(args) == 0:
        return 0  # คืนค่า 0 เพื่อป้องกัน error

    if mode == "avg":
        # โหมดหาเฉลี่ย: นำผลรวมมาหารจำนวนตัวเลข
        return sum(args) / len(args)

    elif mode == "max":
        # โหมดหาค่าสูงสุด: ใช้ฟังก์ชัน max()
        return max(args)

    else:
        # ถ้า mode ไม่ถูกต้อง คืนค่า None
        return None
    
# เรียกโหมดเฉลี่ย: (10 + 20 + 30) / 3 = 20.0
print(stat_calc("avg", 10, 20, 30))    

# เรียกโหมดเฉลี่ยแต่ไม่ส่งตัวเลข args = ()
# เงื่อนไขแรกจะคืนค่า 0
print(stat_calc("avg"))                

# เรียกโหมดหาค่าสูงสุด: max(5, 9, 1, 7) = 9
print(stat_calc("max", 5, 9, 1, 7)) 


```
