```python
def to_upper(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper 
```

```python
@to_upper
def hello():
    return "hello python"
```

```python
hello()
```

```python
def bracket(func):
    def wrapper():
        result = func()
        return f"[{result}]"
    return wrapper
```

```python
@bracket
def hello():
    return "hello python"
```

```python
hello()
```

```python
@to_upper
@bracket
def hello():
    return "hello python"
```

```python
hello()
```
