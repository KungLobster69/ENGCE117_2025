```python
import tkinter as tk
```

```python
def submit():
    username = user_var.get().strip()
    if accept_boo.get():
        result_label.config(text=f"ลงทะเบียนสำเร็จ: {username}",)
    else:
        result_label.config(text="You must accept terms")
```

```python
root = tk.Tk()
user_var = tk.StringVar()
accept_boo = tk.BooleanVar()
root.title("Registration Form")
root.geometry("300x200")
user =tk.Label(root, text="Username")
user.pack()
tk.Entry(root, textvariable=user_var).pack()
tk.Checkbutton(root, text="I Accept Terms",variable=accept_boo).pack()
tk.Button(root, text="Submit",command=submit).pack()
result_label = tk.Label(root, text="")
result_label.pack()
root.mainloop()
```
