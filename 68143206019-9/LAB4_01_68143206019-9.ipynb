```python
def sum_rec(n) :
    if n <= 1 :
        return n
    else :
        return n + sum_rec(n-1)
```

```python
sum_rec(5)
```

```python
def reverse_str(s):
    if len(s) <= 1 :
        return s
    return s[-1] + reverse_str(s[:-1]) 
```

```python
reverse_str("hello")
```

```python
def power(base,exp) :
    if base == 0 :
        return 0
    if exp ==1 :
        return 1
    return power(base,exp-1) * base
```

```python
power(5,3)
```
