```python
class RPGCharacter :
    def __init__(self,n,h,m):
        self.name = n
        self.hp = h
        self.mp = m

    def __add__(self,other):
        if isinstance (other,RPGCharacter) :
            return      RPGCharacter(self.name+"+"+ other.name,
                         self.hp + other.hp,
                         self.mp + other.mp)
        else :
            try :
                int(other)
                self.hp += other
            except ValueError :
                print("Must be RPGCharacter or number")

    def __str__(self):
        return f"{self.name} (HP:{self.hp},MP:{self.mp})"
    
```

```python
Char1 = RPGCharacter("lily",80,50)
Char2 = RPGCharacter("Dahlia",120,40)
```

```python
Char3 = Char1+Char2
```

```python
print(f"{Char3}")
```
