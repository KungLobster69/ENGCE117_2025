```python
import tkinter as tk
```

```python
buttons = [
'7', '8', '9',
'4', '5', '6',
'1', '2', '3',
]
row_val = 1
col_val = 0
```

```python
root = tk.Tk()
root.title("Numpad")

# จอแสดงผล (Display)
display = tk.Entry(root,justify="right")
display.grid(row=0, column=0, columnspan=3, sticky="ew", padx=5, pady=5)

for button in buttons:
    btn = tk.Button(root, text=button, width=5, height=2 ,)
    btn.grid(row=row_val, column=col_val, padx=5, pady=5)
    col_val += 1
    if col_val > 2: # เมื่อถึงคอลัมน์สุดท้าย ให้ขึ้นบรรทัดใหม่
        col_val = 0
        row_val += 1

btn2 =tk.Button(root, text="0", width=15, height=2)
btn2.grid(row=row_val, column=0,columnspan=2, padx=5, pady=5)
btn = tk.Button(root, text=".", width=5, height=2 ,)
btn.grid(row=row_val, column=2, padx=5, pady=5)

root.mainloop()
```
