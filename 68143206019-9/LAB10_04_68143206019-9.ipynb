```python
import tkinter as tk
```

```python
buttons = [
'7', '8', '9',
'4', '5', '6',
'1', '2', '3',
]
row_val = 1
col_val = 0
```

```python
def set_val(val):
    display.insert(tk.END, val)
```

```python
root = tk.Tk()
root.title("Numpad")

# จอแสดงผล (Display)
display = tk.Entry(root,justify="right")
display.grid(row=0, column=0, columnspan=3, sticky="ew", padx=5, pady=5)

for button in buttons:
    btn = tk.Button(root, text=button, width=5, height=2 , command=lambda b=button: set_val(b))
    btn.grid(row=row_val, column=col_val, padx=5, pady=5)
    col_val += 1
    if col_val > 2:
        col_val = 0
        row_val += 1

btn2 =tk.Button(root, text="0", width=15, height=2, command=lambda b=0: set_val(b))
btn2.grid(row=row_val, column=0,columnspan=2, padx=5, pady=5)
btn = tk.Button(root, text=".", width=5, height=2, command=lambda b=".": set_val(b))
btn.grid(row=row_val, column=2, padx=5, pady=5)


root.mainloop()
```
