```python
raw_names = [" somchai ","alice","BOB","c"]
```

```python
cooked = list(map(str.strip, raw_names))
cookeded = list(map(str.lower, cooked))
print(cookeded)
```

```python
cookededed = list(filter(lambda x :len(x)>2, cookeded))
print(cookededed)
```

```python
cookededed = [name.lower().strip() for name in raw_names if len(name) > 2]
print(cookededed)
```
