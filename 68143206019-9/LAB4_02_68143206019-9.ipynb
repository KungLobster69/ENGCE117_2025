```python
def even_gen(limit):
    for i in range (limit+1) :
        if i % 2 == 1 :
            continue
        elif i % 2 == 0 :
            yield i
```

```python
n = even_gen(19)
print(next(n))
print(next(n))
```

```python
def intermidate(start,stop,step) :
    while start < step+stop :
        yield start
        start += step
```

```python
for i in intermidate(0,5,0.5) :
    print(i)
```
