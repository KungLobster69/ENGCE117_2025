```python
class Animal :
    def make_sound(self) :
        print("...(Silence)")
```

```python
class Dog(Animal) :
    def make_sound(self):
        print("Woof! Woof!")
```

```python
class Cat(Animal) :
    def make_sound(self):
        print("Meow~ Meow~")
```

```python
zoo = [Dog(),Cat(),Dog(),Animal()]
for animal in zoo :
    animal.make_sound()
```
