```python
from abc import ABC , abstractmethod

class IoTDevice(ABC) :
    @abstractmethod
    def turn_on(self) :
        pass
    @abstractmethod
    def turn_off(self):
        pass
```

```python
class SmartLight(IoTDevice) :
    def turn_on(self) :
        print("üí° ‡πÑ‡∏ü‡πÄ‡∏õ‡∏¥‡∏î")
    def turn_off(self):
        print("üåë ‡πÑ‡∏ü‡∏õ‡∏¥‡∏î")
```

```python
sl = SmartLight()
```

```python
def run(self):
    self.turn_on()
    self.turn_off()
```

```python
class SmartSpeaker(IoTDevice) :
    def turn_on(self) :
        print("üîä ‡∏•‡∏≥‡πÇ‡∏û‡∏á‡πÄ‡∏õ‡∏¥‡∏î")
```

```python
try :
    run(SmartSpeaker)
except TypeError :
    print("‚ùå Error: Can't instantiate abstract class SmartSpeaker with abstract method turn_off")
```

```python
def test_device(device) :
    if isinstance(device,IoTDevice):
        run(device)
    else:
        print("‚ö†Ô∏è ‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡∏≠‡∏∏‡∏õ‡∏Å‡∏£‡∏ì‡πå IoT ‡∏ó‡∏µ‡πà‡∏£‡∏≠‡∏á‡∏£‡∏±‡∏ö")
```
