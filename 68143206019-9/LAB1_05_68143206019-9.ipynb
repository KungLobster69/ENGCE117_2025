```python
names = ["alice","bob","charlie"]
```

```python
upper = [word.upper() for word in names]
print(f"Upper: {upper}")
```

```python
prices = [100,250,80,500,120]
```

```python
plus_vat = [x for x in prices if x > 200]
vat = [v * 1.07 for v in plus_vat ]
print(f"VAT Prices: {vat}")
```

```python
currencies = {'USD': 35, 'EUR':38,'JPY':0.25}
inverted = {v:k for k,v in currencies.items() if v > 1}
print(inverted)
```
