```python
class playlist :
    def __init__(self,L):
        self.list = list(L)
    
    def __len__(self):
        return len(self.list)
    
    def __getitem__(self,position):
        if isinstance(position,slice):
            return playlist(self.list[position])
        else :
            return self.list[position]

    def __setitem__(self,position,song):
        self.list[position] = song
    
    def __contain__(self,find):
        return find in self.list
    
    def __str__(self):
        return f"{self.list}"
```

```python
s = playlist([1,2,3,4,5])
```

```python
print(s[1])
print(f"New playlist object{s[1:4]}")
```
