```python
def super_sum(*nums) :
    Sum = 0
    for i in nums :
        Sum += i
    print(f"Sum = {Sum}")
```

```python
super_sum(89,48,57)
```

```python
def flexi_cals(mode,*nums) :
    if mode == "add" :
        Sum = 0
        for item in nums :
            Sum = Sum + item
        return Sum
    elif mode == "mult" :
        Sum = 1
        for item in nums :
            Sum *= item
        return Sum
    else : 
        print("Mode Error")
```

```python
Sum = flexi_cals("add",84,5)
print(Sum)
```

```python
def stats(mode,*nums) :
    if mode == "avg" :
        if not nums :
            return 0
        else :
            avg = sum(nums) / len(nums)
        return avg
    elif mode == "max" :
        if not nums :
            return 0
        else :
            Max = max(nums)
        return Max
    else : 
        print("Mode Error")
```

```python
print(stats("avg",15,2,6))
```
