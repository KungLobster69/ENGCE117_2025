```python
class Square :
    def __init__(self,L):
        self.__side = L
    
    @property
    def side(self):
        return self.__side
    
    @side.setter
    def side(self,value) :
        if value > 0 :
            self.__side = value
        else :
            raise ValueError ("Error nuber should more than 0")
    
    @property 
    def area(self) :
        return self.__side * self.__side
```

```python
s = Square(4)
print(f"Side:{s.side},Area:{s.area}")
```

```python
s.side = 5
print(f"New Area: {s.area}")
```

```python
s.side = -10
```
