```python
inventory = {'apple':10,
             'banana':20}
```

```python
def check_price(item_name) :
    Value = inventory.get(item_name,'0')
    return Value
```

```python
search = str(input("name"))
Val = check_price(search)
print(f"Check {search}: {Val}")
```

```python

def add_stock(item,quantity) :
    global inventory
    check = inventory.get(item)
    if check == None :
        update = {item:quantity}
        inventory.update(update)
    else :
        inventory[item] = inventory.get(item,0) + quantity
    print(inventory)      
```

```python
name = str(input("name"))
quantity = int(input("quantity"))
add_stock(name,quantity)
```

```python
stock2 = {'banana':5,'orange':10}
```

```python
for i in stock2 :
    check = inventory.get(i)
    if check == None :
        update = {i:stock2[i]}
        inventory.update(update)
    else :
        inventory[i] = inventory.get(i,0) + stock2[i]
print(inventory)      
```
